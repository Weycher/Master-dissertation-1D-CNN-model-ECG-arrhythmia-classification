```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-
"""
ECG Arrhythmia — MIT-BIH → single-beat .npy → 1-D CNN (PyTorch)
路径已按你的目录固定：
  • 原始记录  /Users/gaochenxi/datasets/mitbih/mitdb/1.0.0/
  • 训练/测试 /Users/gaochenxi/ecg/{train|test}/
如需调整，只改常量区即可。
"""
%matplotlib inline

import pickle, warnings
from pathlib import Path

import wfdb, numpy as np, neurokit2 as nk
import torch
from IPython import display
from torch import nn
from torch.utils import data
from sklearn.preprocessing import StandardScaler
from tqdm import tqdm
from d2l import torch as d2l

# ===================== 常量 ===================== #
ROOT        = Path("/Users/gaochenxi/datasets/mitbih")
RAW_DIR     = Path("/Users/gaochenxi/datasets/mitbih")     # wfdb 下载到此
MITDB_DIR   = RAW_DIR                                       # 48 条记录目录
TRAIN_DIR   = ROOT / "train"                                # 训练拍 .npy
TEST_DIR    = ROOT / "test"                                 # 测试拍 .npy
LABELS_PKL  = ROOT / "labels.pkl"

SIGLEN   = 256   # 单拍长度 (samples)
LEFT_PAD  = 80    # R 峰距窗口左端

TEST_RECS = {str(r) for r in [101,106,108,109,112,114,115,116,118,119,
                              122,124,201,203,205,207,208,209,215,220,223,230,234]}
MIT2AAMI  = {'N':0,'L':0,'R':0,'e':0,'j':0,
             'V':1,'E':1,
             'A':2,'a':2,'J':2,'S':2,
             'F':3,
             'Q':4,'|':4,'f':4}

# ===================== 1. 下载 ===================== #

def download_mitbih():
    """确保 /Users/gaochenxi/datasets/mitbih/mitdb/1.0.0 下 48 条记录齐全"""
    dest = RAW_DIR            # /Users/gaochenxi/datasets/mitbih
    core = MITDB_DIR          # /Users/gaochenxi/datasets/mitbih/mitdb/1.0.0

    need_redownload = (
        not core.exists() or
        len(list(core.glob("*.dat"))) < 48 or
        len(list(core.glob("*.hea"))) < 48 or
        len(list(core.glob("*.atr"))) < 48
    )

    if not need_redownload:
        print("✔ MIT-BIH 已完整下载，跳过")
        return

    print("▶ (重新) 下载 MIT-BIH →", dest)
    # 如已存在残缺目录，先删除干净再下载
    if core.exists():
        import shutil
        shutil.rmtree(core.parent, ignore_errors=True)

    dest.mkdir(parents=True, exist_ok=True)
    wfdb.dl_database("mitdb", dl_dir=str(dest))   # 自动生成 mitdb/1.0.0
    print("✔ 下载完成")
# ===================== 2. 切单拍 ===================== #

def cut_beats():
    """把 .dat 切成单拍 .npy，Z-score 标准化后存 train/ test"""
    if TRAIN_DIR.exists() and any(TRAIN_DIR.glob("*.npy")):
        print("✔ 已有 .npy，跳过 cut_beats()")
        return

    TRAIN_DIR.mkdir(parents=True, exist_ok=True)
    TEST_DIR.mkdir(parents=True,  exist_ok=True)

    hea_files = sorted(MITDB_DIR.glob("*.hea"))
    if len(hea_files) != 48:
        raise FileNotFoundError("MIT-BIH 记录不全，请检查下载目录")

    scaler = StandardScaler()
    print("▶ Pass-1 计算均值/方差")
    for hea in tqdm(hea_files):
        sig, _ = wfdb.rdsamp(str(hea.with_suffix("")), channels=[0])
        sig = nk.signal_filter(sig.squeeze(), sampling_rate=360, lowcut=0.5, highcut=40)
        scaler.partial_fit(sig.reshape(-1, 1))

    print("▶ Pass-2 切单拍并保存 .npy")
    labels = {}
    for hea in tqdm(hea_files):
        rec = hea.stem
        record = wfdb.rdrecord(str(hea.with_suffix("")), channels=[0])
        ann    = wfdb.rdann  (str(hea.with_suffix("")), "atr")

        sig = nk.signal_filter(record.p_signal.squeeze().astype("float32"), sampling_rate=360, lowcut=0.5, highcut=40)
        sig = scaler.transform(sig.reshape(-1,1)).flatten()

        for idx, (r, sym) in enumerate(zip(ann.sample, ann.symbol)):
            if sym not in MIT2AAMI:
                continue
            start, end = r-LEFT_PAD, r-LEFT_PAD+SIGLEN
            if start < 0 or end > len(sig):
                continue
            beat = sig[start:end]
            fname = f"{rec}_{idx:05d}.npy"
            target = TEST_DIR if rec in TEST_RECS else TRAIN_DIR
            np.save(target/fname, beat)
            labels[f"{rec}_{idx:05d}"] = MIT2AAMI[sym]

    with open(LABELS_PKL, "wb") as f:
        pickle.dump(labels, f)

    print(f"✔ 保存单拍: train={len(list(TRAIN_DIR.glob('*.npy')))}, "
          f"test={len(list(TEST_DIR.glob('*.npy')))}")

# ===================== 3. Dataset ===================== #
class ECGDataset(data.Dataset):
    def __init__(self, train=True):
        self.dir   = TRAIN_DIR if train else TEST_DIR
        self.files = sorted(self.dir.glob("*.npy"))
        with open(LABELS_PKL, "rb") as f:
            self.labels = pickle.load(f)

    def __len__(self):
        return len(self.files)

    def __getitem__(self, idx):
        fp = self.files[idx]
        x  = np.load(fp).astype("float32")
        y  = self.labels[fp.stem]
        return torch.tensor(x), torch.tensor(y, dtype=torch.long)


def load_data_ecg(batch_size=256):
    train_set = ECGDataset(True)
    test_set  = ECGDataset(False)
    train_iter = data.DataLoader(train_set, batch_size,
                                 shuffle=True,  num_workers=0, drop_last=True)
    test_iter  = data.DataLoader(test_set,  batch_size,
                                 shuffle=False, num_workers=0)
    return train_iter, test_iter, 5

# ===================== 4. 模型 ===================== #
class ECGNet(nn.Module):
    def __init__(self, classes=5):
        super().__init__()
        self.net = nn.Sequential(
            nn.Conv1d(1, 32, 7, 2, 3), nn.BatchNorm1d(32), nn.ReLU(),
            nn.Conv1d(32, 64, 5, 2, 2), nn.BatchNorm1d(64), nn.ReLU(),
            nn.Conv1d(64,128,3, 2, 1), nn.BatchNorm1d(128), nn.ReLU(),
            nn.AdaptiveAvgPool1d(1), nn.Flatten(),
            nn.Linear(128, classes)
        )
    def forward(self, x):
        return self.net(x.unsqueeze(1))

def init_weights(m):
    if isinstance(m, (nn.Conv1d, nn.Linear)):
        nn.init.kaiming_normal_(m.weight, nonlinearity='relu')

# ===================== 5. 训练 ===================== #
def accuracy(y_hat, y):
    return (y_hat.argmax(1) == y).sum()

def train(num_epochs=15, batch=256, lr=1e-3, device=None):
    device = device or ("mps" if torch.backends.mps.is_available() else
                        "cuda" if torch.cuda.is_available() else "cpu")
    train_iter, test_iter, C = load_data_ecg(batch)
    net = ECGNet(C).to(device);
    net.apply(init_weights)

    loss_fn = nn.CrossEntropyLoss(); opt = torch.optim.Adam(net.parameters(), lr)
    animator = d2l.Animator(xlabel='epoch', legend=['train loss','train acc','test acc'], xlim=[1, num_epochs])
    timer = d2l.Timer()

    for epoch in range(num_epochs):
        metric = d2l.Accumulator(3)
        net.train()
        for X, y in train_iter:
            X, y = X.to(device), y.to(device)
            l = loss_fn(net(X), y)
            opt.zero_grad(); l.backward(); opt.step()
            metric.add(l.item()*y.numel(), accuracy(net(X), y).item(), y.numel())

        train_loss, train_acc = metric[0]/metric[2], metric[1]/metric[2]
        net.eval(); test_metric = d2l.Accumulator(2)
        with torch.no_grad():
            for X, y in test_iter:
                X, y = X.to(device), y.to(device)
                test_metric.add(accuracy(net(X), y).item(), y.numel())
        test_acc = test_metric[0]/test_metric[1]
        animator.add(epoch+1, (train_loss, train_acc, test_acc))
        print(f"epoch {epoch+1:2d}: loss {train_loss:.4f}  train_acc {train_acc:.3f}  test_acc {test_acc:.3f}")
    print(f"⏱  {timer.stop():.1f}s on {device}")

# ===================== 主流程 ===================== #
if __name__ == "__main__":
    warnings.filterwarnings("ignore")
    download_mitbih()
    cut_beats()
    train()
```

```python
import wfdb, matplotlib.pyplot as plt
from collections import Counter

RAW_DIR = Path("/Users/gaochenxi/datasets/mitbih")   # 原始 .dat/.hea/.atr

def plot_record(rec_id="100", secs=10):
    """
    rec_id : '100'‒'234' 字符串
    secs   : 连续可视化多少秒（从第 0 秒开始）
    """
    base = RAW_DIR / rec_id
    record = wfdb.rdrecord(str(base), channels=[0])
    ann    = wfdb.rdann  (str(base), "atr")

    fs   = int(record.fs)                 # 360
    N    = secs * fs
    sig  = record.p_signal[:N,0]          # 取前 secs 秒
    t    = np.arange(N)/fs

    plt.figure(figsize=(12,3))
    plt.plot(t, sig, lw=0.8, label="Lead II")

    # 标注该窗口内的所有 R-peak
    for s, sym in zip(ann.sample, ann.symbol):
        if s < N:
            plt.axvline(s/fs, color='r', ls='--', alpha=.4)
            plt.text(s/fs, sig.max()*0.8, sym, rotation=90, va='bottom', ha='center')
    plt.title(f"Record {rec_id} — first {secs}s   (R-peaks & MIT symbols)")
    plt.xlabel("Time [s]"); plt.ylabel("mV")
    plt.grid(alpha=.3)
    plt.show()

# ▷ 用法：
# plot_record("108", secs=8)
```

```python
# ========= 0 依赖 =============
import wfdb, neurokit2 as nk, numpy as np, matplotlib.pyplot as plt
from pathlib import Path
import matplotlib.pyplot as plt

# ================= 全局绘图样式 ================= #
plt.rcParams['font.family'] = 'Arial'   # 字体
plt.rcParams['font.size'] = 11          # 字号
plt.rcParams['axes.titlesize'] = 11
plt.rcParams['axes.labelsize'] = 11
plt.rcParams['xtick.labelsize'] = 11
plt.rcParams['ytick.labelsize'] = 11
plt.rcParams['legend.fontsize'] = 11
# ========= 1 路径与参数 ========
RAW_DIR = Path("/Users/gaochenxi/datasets/mitbih")  # 原始 .dat/.hea/.atr 存放处
REC_ID  = "234"   # 换成 100–234 中任何一条
sec     = 10      # 查看前多少秒

# ========= 2 读取波形 ===========
rec   = wfdb.rdrecord(str(RAW_DIR/REC_ID), channels=[0])
sig   = rec.p_signal.squeeze().astype("float32")[:sec * rec.fs]  # 截前 sec 秒
fs    = int(rec.fs)                                              # 360

# ========= 3 R-峰 & P/Q/R/S/T ===
_, rpeaks = nk.ecg_peaks(sig, sampling_rate=fs)
_, waves  = nk.ecg_delineate(sig, rpeaks["ECG_R_Peaks"],
                             sampling_rate=fs, method="dwt",
                             show=False, check=False)

# ========= 4 画图 ===============
colors = dict(P="tab:green", Q="tab:purple",
              R="red",       S="tab:orange", T="tab:blue")

plt.figure(figsize=(12,3))
plt.plot(sig, lw=.7, label="Raw ECG")

# ---- R ---------------
r_idx = rpeaks["ECG_R_Peaks"]
r_idx = r_idx[r_idx < len(sig)]
plt.scatter(r_idx, sig[r_idx], color=colors["R"], s=18, marker="^", label="R")

# ---- P/Q/S/T ----------
for key,label in [("ECG_P_Peaks","P"),
                  ("ECG_Q_Peaks","Q"),
                  ("ECG_S_Peaks","S"),
                  ("ECG_T_Peaks","T")]:
    idx = np.asarray(waves.get(key, []), float)
    idx = idx[np.isfinite(idx)].astype(int)
    idx = idx[idx < len(sig)]
    if idx.size:
        plt.scatter(idx, sig[idx], color=colors[label], s=18, label=label)

plt.xlim(0, len(sig))
plt.xlabel(f"Sample (0–{len(sig)-1})  @{fs} Hz")
plt.ylabel(f"mV")
plt.title(f"Record {REC_ID} — first {sec}s with P/Q/R/S/T waves\n\n")
plt.legend(ncol=6, bbox_to_anchor=(1.02,1.15))
plt.savefig("ECG analysis.png", dpi=300, bbox_inches='tight')
plt.tight_layout(); plt.show()
```

```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-
"""
ECG Arrhythmia  –  MIT-BIH → single-beat .npy → 1-D CNN  (PyTorch)
HPO: LocalBackend  +  SingleFidelityScheduler  +  Tuner
(适用于 Syne-Tune 0.14.x)
"""
# ------------------------------------------------------------
# 0. HPO 相关最前置设置
# ------------------------------------------------------------
import os, tempfile, textwrap, json, random, pickle, warnings
os.environ["SYNETUNE_DISABLE_BOTORCH"] = "1"          # ← 如无需 GP-Bayes，关闭 botorch 依赖

from syne_tune.config_space import uniform, randint, choice
from syne_tune.backend.local_backend import LocalBackend
from syne_tune.optimizer.schedulers.single_fidelity_scheduler import (SingleFidelityScheduler)
from syne_tune import Tuner, Reporter
from syne_tune import StoppingCriterion

from pathlib import Path
import wfdb, neurokit2 as nk, numpy as np, torch
from torch import nn
from torch.utils import data
from sklearn.preprocessing import StandardScaler
from tqdm import tqdm
import string, tempfile, textwrap

# ---------- 常量 ----------
ROOT = Path("/Users/gaochenxi/datasets/mitbih")
RAW_DIR  = ROOT
MITDB_DIR = RAW_DIR                    # ★ 补：download_mitbih() 要用
TRAIN_DIR, TEST_DIR = ROOT/"train", ROOT/"test"
LABELS_PKL = ROOT/"labels.pkl"
SIGLEN, LEFT_PAD = 256, 80
TEST_RECS = {str(r) for r in [101,106,108,109,112,114,115,116,118,119,
                              122,124,201,203,205,207,208,209,215,220,223,230,234]}
MIT2AAMI  = {'N':0,'L':0,'R':0,'e':0,'j':0,'V':1,'E':1,
             'A':2,'a':2,'J':2,'S':2,'F':3,'Q':4,'|':4,'f':4}

# ===================== 1. 下载 ===================== #

def download_mitbih():
    """确保 /Users/gaochenxi/datasets/mitbih/mitdb/1.0.0 下 48 条记录齐全"""
    dest = RAW_DIR            # /Users/gaochenxi/datasets/mitbih
    core = MITDB_DIR          # /Users/gaochenxi/datasets/mitbih/mitdb/1.0.0

    need_redownload = (
        not core.exists() or
        len(list(core.glob("*.dat"))) < 48 or
        len(list(core.glob("*.hea"))) < 48 or
        len(list(core.glob("*.atr"))) < 48
    )

    if not need_redownload:
        print("✔ MIT-BIH 已完整下载，跳过")
        return

    print("▶ (重新) 下载 MIT-BIH →", dest)
    # 如已存在残缺目录，先删除干净再下载
    if core.exists():
        import shutil
        shutil.rmtree(core.parent, ignore_errors=True)

    dest.mkdir(parents=True, exist_ok=True)
    wfdb.dl_database("mitdb", dl_dir=str(dest))   # 自动生成 mitdb/1.0.0
    print("✔ 下载完成")
# ===================== 2. 切单拍 ===================== #

def cut_beats():
    """把 .dat 切成单拍 .npy，Z-score 标准化后存 train/ test"""
    if TRAIN_DIR.exists() and any(TRAIN_DIR.glob("*.npy")):
        print("✔ 已有 .npy，跳过 cut_beats()")
        return

    TRAIN_DIR.mkdir(parents=True, exist_ok=True)
    TEST_DIR.mkdir(parents=True,  exist_ok=True)

    hea_files = sorted(MITDB_DIR.glob("*.hea"))
    if len(hea_files) != 48:
        raise FileNotFoundError("MIT-BIH 记录不全，请检查下载目录")

    scaler = StandardScaler()
    print("▶ Pass-1 计算均值/方差")
    for hea in tqdm(hea_files):
        sig, _ = wfdb.rdsamp(str(hea.with_suffix("")), channels=[0])
        sig = nk.signal_filter(sig.squeeze(), sampling_rate=360, lowcut=0.5, highcut=40)
        scaler.partial_fit(sig.reshape(-1, 1))

    print("▶ Pass-2 切单拍并保存 .npy")
    labels = {}
    for hea in tqdm(hea_files):
        rec = hea.stem
        record = wfdb.rdrecord(str(hea.with_suffix("")), channels=[0])
        ann    = wfdb.rdann  (str(hea.with_suffix("")), "atr")

        sig = nk.signal_filter(record.p_signal.squeeze().astype("float32"), sampling_rate=360, lowcut=0.5, highcut=40)
        sig = scaler.transform(sig.reshape(-1,1)).flatten()

        for idx, (r, sym) in enumerate(zip(ann.sample, ann.symbol)):
            if sym not in MIT2AAMI:
                continue
            start, end = r-LEFT_PAD, r-LEFT_PAD+SIGLEN
            if start < 0 or end > len(sig):
                continue
            beat = sig[start:end]
            fname = f"{rec}_{idx:05d}.npy"
            target = TEST_DIR if rec in TEST_RECS else TRAIN_DIR
            np.save(target/fname, beat)
            labels[f"{rec}_{idx:05d}"] = MIT2AAMI[sym]

    with open(LABELS_PKL, "wb") as f:
        pickle.dump(labels, f)

    print(f"✔ 保存单拍: train={len(list(TRAIN_DIR.glob('*.npy')))}, "
          f"test={len(list(TEST_DIR.glob('*.npy')))}")

# ===================== 3. Dataset ===================== #
class ECGDataset(data.Dataset):
    def __init__(self, train=True):
        self.dir = TRAIN_DIR if train else TEST_DIR
        self.files = sorted(self.dir.glob("*.npy"))
        with open(LABELS_PKL, "rb") as f:
            self.labels = pickle.load(f)
    def __len__(self):  return len(self.files)
    def __getitem__(self, idx):
        fp = self.files[idx]
        x = np.load(fp).astype("float32")
        y = self.labels[fp.stem]
        return torch.tensor(x), torch.tensor(y, dtype=torch.long)

def load_data_ecg(batch):
    train_set, test_set = ECGDataset(True), ECGDataset(False)
    train_iter = data.DataLoader(train_set, batch, shuffle=True,  num_workers=0, drop_last=True)
    test_iter  = data.DataLoader(test_set,  batch, shuffle=False, num_workers=0)
    return train_iter, test_iter

# ---------- 网络骨架 ----------
def build_net(c1, c2, c3):
    return nn.Sequential(
        nn.Conv1d(1, c1, 7, 2, 3), nn.BatchNorm1d(c1), nn.ReLU(),
        nn.Conv1d(c1, c2, 5, 2, 2), nn.BatchNorm1d(c2), nn.ReLU(),
        nn.Conv1d(c2, c3, 3, 2, 1), nn.BatchNorm1d(c3), nn.ReLU(),
        nn.AdaptiveAvgPool1d(1), nn.Flatten(),
        nn.Linear(c3, 5)
    )

# ------------------------------------------------------------
# 2. 把 train_fn 写成临时脚本，保存到磁盘
# ------------------------------------------------------------

# ------------------------------------------------------------
# 3. 搜索空间 + Backend + Scheduler + Tuner
# ------------------------------------------------------------

# ------------------------------------------------------------
# 4. 主流程
# ------------------------------------------------------------
```

```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-
"""
ECG Arrhythmia Classification — MIT-BIH → Multi-lead + FFT + Down-sampled 1-D CNN (PyTorch)
-------------------------------------------------------------------------------------------------
增强特性：
  • **多导联 (MLII + V1)** 2 路时域
  • **FFT 幅度谱** 2 路频域 → 共 4 通道
  • **降采样** 360 Hz → 180 Hz

新增：**在训练日志与动画中实时输出 *macro-F1*（测试集）**，便于不均衡类别评估。
"""

import pickle, warnings, shutil
from pathlib import Path
from typing import List

import wfdb, numpy as np, neurokit2 as nk
import torch
from torch import nn
from torch.utils import data
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import f1_score
from tqdm import tqdm
from d2l import torch as d2l

# ===================== 常量 ===================== #
ROOT   = Path("/Users/gaochenxi/datasets/mitbih")  # ★ 按需修改
RAW_DIR = ROOT
MITDB_DIR = RAW_DIR / "mitdb" / "1.0.0"          # 若文件就在 mitdb/，把路径改掉即可
TRAIN_DIR = ROOT / "train"
TEST_DIR  = ROOT / "test"
LABELS_PKL = ROOT / "labels.pkl"

LEADS      = [0, 1]   # MLII, V1
ORIG_FS    = 360
TARGET_FS  = 180
SIGLEN     = 256
LEFT_PAD   = 80

TEST_RECS = {str(r) for r in [101,106,108,109,112,114,115,116,118,119,
                              122,124,201,203,205,207,208,209,215,220,223,230,234]}
MIT2AAMI  = {'N':0,'L':0,'R':0,'e':0,'j':0,
             'V':1,'E':1,
             'A':2,'a':2,'J':2,'S':2,
             'F':3,
             'Q':4,'|':4,'f':4}

LEAD_N       = len(LEADS)          # 2
IN_CHANNELS  = LEAD_N * 2          # 4 通道
AUG_FACTOR   = ORIG_FS / TARGET_FS
LEFT_PAD_RAW = int(LEFT_PAD * AUG_FACTOR)
SIGLEN_RAW   = int(SIGLEN * AUG_FACTOR)
FORCE_REBUILD = False

# ===================== 1. 下载 ===================== #

def download_mitbih() -> None:
    ok = (
        MITDB_DIR.exists() and
        len(list(MITDB_DIR.glob('*.hea'))) == 48 and
        len(list(MITDB_DIR.glob('*.dat'))) == 48 and
        len(list(MITDB_DIR.glob('*.atr'))) == 48
    )
    if ok:
        print('✔ MIT-BIH 已完整下载，跳过')
        return
    print('▶ (重新) 下载 MIT-BIH →', RAW_DIR)
    shutil.rmtree(MITDB_DIR.parent, ignore_errors=True)
    RAW_DIR.mkdir(parents=True, exist_ok=True)
    wfdb.dl_database('mitdb', dl_dir=str(RAW_DIR))
    print('✔ 下载完成')

# ===================== 2. 切单拍 ===================== #

def _filter(sig: np.ndarray, fs: int = ORIG_FS) -> np.ndarray:
    return nk.signal_filter(sig, sampling_rate=fs, lowcut=0.5, highcut=40)

def _resample(sig: np.ndarray) -> np.ndarray:
    return nk.signal_resample(sig, desired_length=SIGLEN)

def _already_ok(dir_: Path) -> bool:
    samp = list(dir_.glob('*.npy'))[:5]
    if not samp:
        return False
    return all(np.load(f, mmap_mode='r').shape == (LEAD_N, SIGLEN) for f in samp)

def cut_beats() -> None:
    if not FORCE_REBUILD and TRAIN_DIR.exists() and _already_ok(TRAIN_DIR):
        print('✔ 已有多导联 .npy，跳过 cut_beats()')
        return
    shutil.rmtree(TRAIN_DIR, ignore_errors=True)
    shutil.rmtree(TEST_DIR,  ignore_errors=True)
    TRAIN_DIR.mkdir(parents=True, exist_ok=True)
    TEST_DIR.mkdir(parents=True,  exist_ok=True)

    hea_files = sorted(MITDB_DIR.glob('*.hea'))
    if len(hea_files) != 48:
        raise FileNotFoundError('MIT-BIH 记录不全，请检查下载目录')

    scaler = StandardScaler()
    print('▶ Pass-1 计算全局均值/方差')
    for hea in tqdm(hea_files):
        sig, _ = wfdb.rdsamp(str(hea.with_suffix('')), channels=LEADS)
        sig = np.vstack([_filter(sig[:, i]) for i in range(LEAD_N)]).T
        scaler.partial_fit(sig.reshape(-1,1))

    print('▶ Pass-2 切单拍并保存')
    labels = {}
    for hea in tqdm(hea_files):
        rec = hea.stem
        record = wfdb.rdrecord(str(hea.with_suffix('')), channels=LEADS)
        ann    = wfdb.rdann  (str(hea.with_suffix('')), 'atr')
        sig = record.p_signal.astype('float32')
        sig = np.vstack([_filter(sig[:, i]) for i in range(LEAD_N)]).T
        sig = scaler.transform(sig.reshape(-1,1)).reshape(-1, LEAD_N)
        for idx, (r, sym) in enumerate(zip(ann.sample, ann.symbol)):
            if sym not in MIT2AAMI:
                continue
            start, end = r-LEFT_PAD_RAW, r-LEFT_PAD_RAW+SIGLEN_RAW
            if start < 0 or end > len(sig):
                continue
            beat = sig[start:end].T
            beat = np.stack([_resample(ch) for ch in beat])
            np.save((TEST_DIR if rec in TEST_RECS else TRAIN_DIR)/f"{rec}_{idx:05d}.npy", beat)
            labels[f"{rec}_{idx:05d}"] = MIT2AAMI[sym]
    with open(LABELS_PKL,'wb') as f:
        pickle.dump(labels, f)
    print('✔ 切片完成')

# ===================== 3. Dataset ===================== #
class ECGDataset(data.Dataset):
    def __init__(self, train: bool=True):
        self.dir   = TRAIN_DIR if train else TEST_DIR
        self.files = [f for f in sorted(self.dir.glob('*.npy')) if np.load(f, mmap_mode='r').shape==(LEAD_N, SIGLEN)]
        with open(LABELS_PKL,'rb') as f:
            self.labels = pickle.load(f)
    def __len__(self):
        return len(self.files)
    def __getitem__(self, idx):
        fp = self.files[idx]
        x_td = np.load(fp)
        x_fd = np.abs(np.fft.fft(x_td, axis=1)).astype('float32')
        x = np.concatenate([x_td, x_fd], axis=0)
        y = self.labels[fp.stem]
        return torch.tensor(x), torch.tensor(y)

def load_data_ecg(batch_size=256):
    train_set, test_set = ECGDataset(True), ECGDataset(False)
    C = len(set(train_set.labels.values()))
    return (
        data.DataLoader(train_set, batch_size, shuffle=True, num_workers=0, drop_last=True),
        data.DataLoader(test_set,  batch_size, shuffle=False, num_workers=0),
        C
    )

# ===================== 4. 模型 ===================== #
class ECGNet(nn.Module):
    def __init__(self, in_channels=IN_CHANNELS, classes=5):
        super().__init__()
        self.net = nn.Sequential(
            nn.Conv1d(in_channels,32,7,2,3), nn.BatchNorm1d(32), nn.ReLU(),
            nn.Conv1d(32,64,5,2,2),          nn.BatchNorm1d(64), nn.ReLU(),
            nn.Conv1d(64,128,3,2,1),         nn.BatchNorm1d(128),nn.ReLU(),
            nn.AdaptiveAvgPool1d(1), nn.Flatten(),
            nn.Linear(128, classes)
        )
    def forward(self,x):
        return self.net(x)

def init_weights(m):
    if isinstance(m,(nn.Conv1d,nn.Linear)):
        nn.init.kaiming_normal_(m.weight, nonlinearity='relu')

# ===================== 5. 训练 ===================== #

def accuracy(y_hat,y):
    return (y_hat.argmax(1)==y).sum()

def train(num_epochs=15, batch=256, lr=1e-3, device=None):
    device = device or ('mps' if torch.backends.mps.is_available() else 'cuda' if torch.cuda.is_available() else 'cpu')
    train_iter, test_iter, C = load_data_ecg(batch)
    net = ECGNet(IN_CHANNELS, C).to(device)
    net.apply(init_weights)

    loss_fn, opt = nn.CrossEntropyLoss(), torch.optim.Adam(net.parameters(), lr)
    animator = d2l.Animator(xlabel='epoch', xlim=[1, num_epochs], legend=['train loss','train acc','test acc','test f1'])
    timer = d2l.Timer()

    for epoch in range(num_epochs):
        metric = d2l.Accumulator(3)
        net.train()
        for X, y in train_iter:
            X, y = X.to(device), y.to(device)
            l = loss_fn(net(X), y)
            opt.zero_grad(); l.backward(); opt.step()
            metric.add(l.item()*y.numel(), accuracy(net(X), y).item(), y.numel())
        train_loss, train_acc = metric[0]/metric[2], metric[1]/metric[2]

        net.eval(); test_metric = d2l.Accumulator(2); preds, trues = [], []
        with torch.no_grad():
            for X, y in test_iter:
                X, y = X.to(device), y.to(device)
                y_hat = net(X)
                test_metric.add(accuracy(y_hat, y).item(), y.numel())
                preds.append(y_hat.argmax(1).cpu()); trues.append(y.cpu())
        test_acc = test_metric[0]/test_metric[1]
        preds = torch.cat(preds); trues = torch.cat(trues)
        test_f1 = f1_score(trues, preds, average='macro')

        animator.add(epoch+1, (train_loss, train_acc, test_acc, test_f1))
        print(f"epoch {epoch+1:2d}: loss {train_loss:.4f}  train_acc {train_acc:.3f}  test_acc {test_acc:.3f}  test_f1 {test_f1:.3f}")
    print(f"⏱ {timer.stop():.1f}s on {device}")

# ===================== 主流程 ===================== #
if __name__ == '__main__':
    warnings.filterwarnings('ignore')
    download_mitbih()
    cut_beats()
    train(num_epochs=15)
```

```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-
"""
ECG Arrhythmia Classification — MIT‑BIH → Multi‑lead + FFT + Down‑sampled 1‑D CNN (PyTorch)
---------------------------------------------------------------------
按照《动手学深度学习》李沐老师课件的代码风格，对原单导联示例做了三项核心增强：
  1. **多导联 (MLII + V1)** ‑ 同时使用两路信号作为时域输入。
  2. **FFT 频域特征** ‑ 对每个导联再附加一条幅度谱通道，实现时频融合 (共 4 通道)。
  3. **降采样** ‑ 将原始 360 Hz 重采样到 180 Hz，再裁剪/对齐固定长度 256。

其余训练、评测与可视化部分保持李沐老师在 d2l 中的习惯写法。

路径仍可通过常量区一键修改。
"""

import pickle, warnings, math
from pathlib import Path

import wfdb, numpy as np, neurokit2 as nk
import torch
from torch import nn
from torch.utils import data
from IPython import display  # noqa: F401  (仅 jupyter 可视化时用)
from sklearn.preprocessing import StandardScaler
from tqdm import tqdm
from d2l import torch as d2l

# ===================== 常量 ===================== #
ROOT        = Path("/Users/gaochenxi/datasets/mitbih")
RAW_DIR     = Path("/Users/gaochenxi/datasets/mitbih")     # wfdb 下载到此
MITDB_DIR   = RAW_DIR                                       # 48 条记录目录
TRAIN_DIR   = ROOT / "train"                                # 训练拍 .npy
TEST_DIR    = ROOT / "test"                                 # 测试拍 .npy
LABELS_PKL  = ROOT / "labels.pkl"

# 信号相关
LEADS       = [0, 1]      # MLII, V1
ORIG_FS     = 360         # MIT‑BIH 原采样率
TARGET_FS   = 180         # 降采样目标采样率 (ORIG_FS // 2)
SIGLEN      = 256         # 降采样后固定长度 (samples)
LEFT_PAD    = 80          # R 峰距窗口左端 (TARGET_FS 下)

TEST_RECS = {str(r) for r in [101,106,108,109,112,114,115,116,118,119,
                              122,124,201,203,205,207,208,209,215,220,223,230,234]}
MIT2AAMI  = {'N':0,'L':0,'R':0,'e':0,'j':0,
             'V':1,'E':1,
             'A':2,'a':2,'J':2,'S':2,
             'F':3,
             'Q':4,'|':4,'f':4}

# 方便后续使用的派生量 (无需手改)
LEAD_N        = len(LEADS)          # 2
IN_CHANNELS   = LEAD_N * 2          # 时域 + 频域
AUG_FACTOR    = ORIG_FS / TARGET_FS # 2.0
LEFT_PAD_RAW  = int(LEFT_PAD * AUG_FACTOR)
SIGLEN_RAW    = int(SIGLEN * AUG_FACTOR)

# ===================== 1. 下载 ===================== #

def download_mitbih() -> None:
    """确保 MIT‑BIH 48 条记录在本地"""
    need_redownload = (
        not MITDB_DIR.exists() or
        len(list(MITDB_DIR.glob("*.dat"))) < 48 or
        len(list(MITDB_DIR.glob("*.hea"))) < 48 or
        len(list(MITDB_DIR.glob("*.atr"))) < 48
    )
    if not need_redownload:
        print("✔ MIT‑BIH 已完整下载，跳过")
        return

    print("▶ (重新) 下载 MIT‑BIH →", RAW_DIR)
    import shutil
    shutil.rmtree(MITDB_DIR.parent, ignore_errors=True)
    RAW_DIR.mkdir(parents=True, exist_ok=True)
    wfdb.dl_database("mitdb", dl_dir=str(RAW_DIR))  # 自动生成 mitdb/1.0.0
    print("✔ 下载完成")

# ===================== 2. 切单拍 ===================== #

def _filter(sig: np.ndarray, fs: int = ORIG_FS) -> np.ndarray:
    """0.5–40 Hz 带通滤波 (neurokit2)"""
    return nk.signal_filter(sig, sampling_rate=fs, lowcut=0.5, highcut=40)

def _resample(sig: np.ndarray, desired_len: int = SIGLEN) -> np.ndarray:
    """重采样至 desired_len 样本 (≈ TARGET_FS)"""
    return nk.signal_resample(sig, desired_length=desired_len)

def cut_beats() -> None:
    """把 .dat 切成多导联单拍 .npy，并做 Z‑score 标准化 (train+test)"""
    if TRAIN_DIR.exists() and any(TRAIN_DIR.glob("*.npy")):
        print("✔ 已有 .npy，跳过 cut_beats()")
        return

    TRAIN_DIR.mkdir(parents=True, exist_ok=True)
    TEST_DIR.mkdir(parents=True,  exist_ok=True)

    hea_files = sorted(MITDB_DIR.glob("*.hea"))
    if len(hea_files) != 48:
        raise FileNotFoundError("MIT‑BIH 记录不全，请检查下载目录")

    scaler = StandardScaler()
    print("▶ Pass‑1 计算均值/方差 (全部导联)")
    for hea in tqdm(hea_files):
        sig, _ = wfdb.rdsamp(str(hea.with_suffix("")), channels=LEADS)
        sig = np.vstack([_filter(sig[:, i]) for i in range(LEAD_N)]).T  # shape (T, L)
        scaler.partial_fit(sig.reshape(-1, 1))

    print("▶ Pass‑2 切单拍并保存 .npy")
    labels = {}
    for hea in tqdm(hea_files):
        rec = hea.stem
        record = wfdb.rdrecord(str(hea.with_suffix("")), channels=LEADS)
        ann    = wfdb.rdann  (str(hea.with_suffix("")), "atr")

        sig = record.p_signal.astype("float32")  # shape (T, L)
        sig = np.vstack([_filter(sig[:, i]) for i in range(LEAD_N)]).T
        sig = scaler.transform(sig.reshape(-1, 1)).reshape(-1, LEAD_N)  # Z‑score

        for idx, (r, sym) in enumerate(zip(ann.sample, ann.symbol)):
            if sym not in MIT2AAMI:
                continue
            start, end = r - LEFT_PAD_RAW, r - LEFT_PAD_RAW + SIGLEN_RAW
            if start < 0 or end > len(sig):
                continue
            beat = sig[start:end]             # shape (SIGLEN_RAW, L)
            beat = beat.T                     # (L, SIGLEN_RAW) — 先转为 (L, T)
            beat = np.stack([_resample(ch, SIGLEN) for ch in beat])  # (L, SIGLEN)
            fname   = f"{rec}_{idx:05d}.npy"
            target  = TEST_DIR if rec in TEST_RECS else TRAIN_DIR
            np.save(target / fname, beat.astype("float32"))
            labels[f"{rec}_{idx:05d}"] = MIT2AAMI[sym]

    with open(LABELS_PKL, "wb") as f:
        pickle.dump(labels, f)

    print(f"✔ 保存单拍: train={len(list(TRAIN_DIR.glob('*.npy')))}, "
          f"test={len(list(TEST_DIR.glob('*.npy')))}")

# ===================== 3. Dataset ===================== #
class ECGDataset(data.Dataset):
    """按需拼接时域 + FFT 通道"""
    def __init__(self, train: bool = True):
        self.dir   = TRAIN_DIR if train else TEST_DIR
        self.files = sorted(self.dir.glob("*.npy"))
        with open(LABELS_PKL, "rb") as f:
            self.labels = pickle.load(f)

    def __len__(self):
        return len(self.files)

    def __getitem__(self, idx):
        fp   = self.files[idx]
        x_td = np.load(fp)                        # (L, SIGLEN)
        # 频域幅度谱 (对称截取同长)
        x_fd = np.abs(np.fft.fft(x_td, axis=1)).astype("float32")
        x    = np.concatenate([x_td, x_fd], axis=0)  # (L*2, SIGLEN)
        y    = self.labels[fp.stem]
        return torch.tensor(x), torch.tensor(y, dtype=torch.long)

def load_data_ecg(batch_size: int = 256):
    train_set  = ECGDataset(True)
    test_set   = ECGDataset(False)
    classes    = len(set(train_set.labels.values()))
    train_iter = data.DataLoader(train_set, batch_size, shuffle=True,  num_workers=0, drop_last=True)
    test_iter  = data.DataLoader(test_set,  batch_size, shuffle=False, num_workers=0)
    return train_iter, test_iter, classes

# ===================== 4. 模型 ===================== #
class ECGNet(nn.Module):
    def __init__(self, in_channels: int = IN_CHANNELS, classes: int = 5):
        super().__init__()
        self.net = nn.Sequential(
            nn.Conv1d(in_channels,  32, 7, 2, 3), nn.BatchNorm1d(32),  nn.ReLU(),
            nn.Conv1d(32,  64, 5, 2, 2),          nn.BatchNorm1d(64),  nn.ReLU(),
            nn.Conv1d(64, 128, 3, 2, 1),          nn.BatchNorm1d(128), nn.ReLU(),
            nn.AdaptiveAvgPool1d(1), nn.Flatten(),
            nn.Linear(128, classes)
        )

    def forward(self, x):
        return self.net(x)  # x: (B, C, T)

def init_weights(m):
    if isinstance(m, (nn.Conv1d, nn.Linear)):
        nn.init.kaiming_normal_(m.weight, nonlinearity="relu")

# ===================== 5. 训练 ===================== #

def accuracy(y_hat, y):
    return (y_hat.argmax(1) == y).sum()

def train(num_epochs: int = 15, batch: int = 256, lr: float = 1e-3, device: str | None = None):
    device = device or (
        "mps"  if torch.backends.mps.is_available() else
        "cuda" if torch.cuda.is_available() else
        "cpu")

    train_iter, test_iter, C = load_data_ecg(batch)
    net = ECGNet(IN_CHANNELS, C).to(device)
    net.apply(init_weights)

    loss_fn  = nn.CrossEntropyLoss()
    opt      = torch.optim.Adam(net.parameters(), lr)
    animator = d2l.Animator(xlabel="epoch", xlim=[1, num_epochs], legend=["train loss", "train acc", "test acc"])
    timer    = d2l.Timer()

    for epoch in range(num_epochs):
        metric = d2l.Accumulator(3)
        net.train()
        for X, y in train_iter:
            X, y = X.to(device), y.to(device)
            l = loss_fn(net(X), y)
            opt.zero_grad(); l.backward(); opt.step()
            metric.add(l.item() * y.numel(), accuracy(net(X), y).item(), y.numel())

        train_loss, train_acc = metric[0] / metric[2], metric[1] / metric[2]
        net.eval(); test_metric = d2l.Accumulator(2)
        with torch.no_grad():
            for X, y in test_iter:
                X, y = X.to(device), y.to(device)
                test_metric.add(accuracy(net(X), y).item(), y.numel())
        test_acc = test_metric[0] / test_metric[1]
        animator.add(epoch + 1, (train_loss, train_acc, test_acc))
        display.clear_output(wait=True)   # ① 先清空，但浏览器“等下一帧”
        display.display(animator.fig)     # ② 重用同一张 fig
        print(f"epoch {epoch + 1:2d}: loss {train_loss:.4f}  train_acc {train_acc:.3f}  test_acc {test_acc:.3f}")
    print(f"⏱  {timer.stop():.1f}s on {device}")

# ===================== 主流程 ===================== #
if __name__ == "__main__":
    warnings.filterwarnings("ignore")
    download_mitbih()
    cut_beats()
    train()
```

```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-
"""
ECG Arrhythmia Classification — MIT-BIH → Multi-lead + FFT + Down-sampled 1-D CNN (PyTorch)
---------------------------------------------------------------------

  • Dropout 正则化（每个 Conv Block + 最后全连接）
  • Adam 权重衰减 (L2) = 1e-4
  • ReduceLROnPlateau 学习率调度 (factor 0.5, patience 2)
  • Early-Stopping (patience 5) — 自动保存最佳模型 best_model.pth
  • 固定随机种子保证可复现

用法：python ecg_arrhythmia_cnn_enhanced.py  # 直接跑完整流程
"""

import pickle, warnings, math, random, time
from pathlib import Path

import wfdb, numpy as np, neurokit2 as nk
import torch
from torch import nn
from torch.utils import data
from IPython import display  
from sklearn.preprocessing import StandardScaler
from tqdm import tqdm
from d2l import torch as d2l
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# ================= 全局绘图样式 ================= #
plt.rcParams['font.family'] = 'Arial'   # 字体
plt.rcParams['font.size'] = 11          # 字号
plt.rcParams['axes.titlesize'] = 11
plt.rcParams['axes.labelsize'] = 11
plt.rcParams['xtick.labelsize'] = 11
plt.rcParams['ytick.labelsize'] = 11
plt.rcParams['legend.fontsize'] = 11

# ===================== 固定随机种子 ===================== #
SEED = 42
random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(SEED)

def seed_worker(_):
    np.random.seed(SEED)
    random.seed(SEED)

# ===================== 常量 ===================== #
ROOT        = Path("/Users/gaochenxi/datasets/mitbih")
RAW_DIR     = ROOT                      # wfdb 下载到此
MITDB_DIR   = RAW_DIR                   # 48 条记录目录
TRAIN_DIR   = ROOT / "train"            # 训练拍 .npy
TEST_DIR    = ROOT / "test"             # 测试拍 .npy
LABELS_PKL  = ROOT / "labels.pkl"

# 信号相关
LEADS       = [0, 1]      # MLII, V5
ORIG_FS     = 360         # MIT-BIH 原采样率
TARGET_FS   = 180         # 降采样目标采样率 (ORIG_FS // 2)
SIGLEN      = 128         # 降采样后固定长度 (samples)
LEFT_PAD    = 40          # R 峰距窗口左端 (TARGET_FS 下)

TEST_RECS = {str(r) for r in [101,106,108,109,112,114,115,116,118,119,
                              122,124,201,203,205,207,208,209,215,220,223,230,234]}
MIT2AAMI  = {'N':0,'L':0,'R':0,'e':0,'j':0,
             'V':1,'E':1,
             'A':2,'a':2,'J':2,'S':2,
             'F':3,
             'Q':4,'|':4,'f':4}


LEAD_N        = len(LEADS)          # 2
IN_CHANNELS   = LEAD_N * 2          # 时域 + 频域
AUG_FACTOR    = ORIG_FS / TARGET_FS # 2.0
LEFT_PAD_RAW  = int(LEFT_PAD * AUG_FACTOR)
SIGLEN_RAW    = int(SIGLEN * AUG_FACTOR)

# ===================== 1. 下载 ===================== #

def download_mitbih() -> None:
    """确保 MIT-BIH 48 条记录在本地"""
    need_redownload = (
        not MITDB_DIR.exists() or
        len(list(MITDB_DIR.glob("*.dat"))) < 48 or
        len(list(MITDB_DIR.glob("*.hea"))) < 48 or
        len(list(MITDB_DIR.glob("*.atr"))) < 48
    )
    if not need_redownload:
        print("✔ MIT-BIH 已完整下载，跳过")
        return

    print("▶ (重新) 下载 MIT-BIH →", RAW_DIR)
    import shutil
    shutil.rmtree(MITDB_DIR.parent, ignore_errors=True)
    RAW_DIR.mkdir(parents=True, exist_ok=True)
    wfdb.dl_database("mitdb", dl_dir=str(RAW_DIR))  # 自动生成 mitdb/1.0.0
    print("✔ 下载完成")

# ===================== 2. 切单拍 ===================== #

def _filter(sig: np.ndarray, fs: int = ORIG_FS) -> np.ndarray:
    """0.5–40 Hz 带通滤波 (neurokit2)"""
    return nk.signal_filter(sig, sampling_rate=fs, lowcut=0.5, highcut=40)


def _resample(sig: np.ndarray, desired_len: int = SIGLEN) -> np.ndarray:
    """重采样至 desired_len 样本 (≈ TARGET_FS)"""
    return nk.signal_resample(sig, desired_length=desired_len)


def cut_beats() -> None:
    """把 .dat 切成多导联单拍 .npy，并做 Z-score 标准化 (train+test)"""
    if TRAIN_DIR.exists() and any(TRAIN_DIR.glob("*.npy")):
        print("✔ 已有 .npy，跳过 cut_beats()")
        return

    TRAIN_DIR.mkdir(parents=True, exist_ok=True)
    TEST_DIR.mkdir(parents=True,  exist_ok=True)

    hea_files = sorted(MITDB_DIR.glob("*.hea"))
    if len(hea_files) != 48:
        raise FileNotFoundError("MIT-BIH 记录不全，请检查下载目录")

    scaler = StandardScaler()
    print("▶ Pass-1 计算均值/方差 (全部导联)")
    for hea in tqdm(hea_files):
        sig, _ = wfdb.rdsamp(str(hea.with_suffix("")), channels=LEADS)
        sig = np.vstack([_filter(sig[:, i]) for i in range(LEAD_N)]).T  # shape (T, L)
        scaler.partial_fit(sig.reshape(-1, 1))

    print("▶ Pass-2 切单拍并保存 .npy")
    labels = {}
    for hea in tqdm(hea_files):
        rec = hea.stem
        record = wfdb.rdrecord(str(hea.with_suffix("")), channels=LEADS)
        ann    = wfdb.rdann  (str(hea.with_suffix("")), "atr")

        sig = record.p_signal.astype("float32")  # shape (T, L)
        sig = np.vstack([_filter(sig[:, i]) for i in range(LEAD_N)]).T
        sig = scaler.transform(sig.reshape(-1, 1)).reshape(-1, LEAD_N)  # Z-score

        for idx, (r, sym) in enumerate(zip(ann.sample, ann.symbol)):
            if sym not in MIT2AAMI:
                continue
            start, end = r - LEFT_PAD_RAW, r - LEFT_PAD_RAW + SIGLEN_RAW
            if start < 0 or end > len(sig):
                continue
            beat = sig[start:end]             # shape (SIGLEN_RAW, L)
            beat = beat.T                     # (L, SIGLEN_RAW)
            beat = np.stack([_resample(ch, SIGLEN) for ch in beat])  # (L, SIGLEN)
            fname   = f"{rec}_{idx:05d}.npy"
            target  = TEST_DIR if rec in TEST_RECS else TRAIN_DIR
            np.save(target / fname, beat.astype("float32"))
            labels[f"{rec}_{idx:05d}"] = MIT2AAMI[sym]

    with open(LABELS_PKL, "wb") as f:
        pickle.dump(labels, f)

    print(f"✔ 保存单拍: train={len(list(TRAIN_DIR.glob('*.npy')))}, "
          f"test={len(list(TEST_DIR.glob('*.npy')))}")

# ===================== 3. Dataset ===================== #
class ECGDataset(data.Dataset):
    """按需拼接时域 + FFT 通道"""
    def __init__(self, train: bool = True):
        self.dir   = TRAIN_DIR if train else TEST_DIR
        self.files = sorted(self.dir.glob("*.npy"))
        with open(LABELS_PKL, "rb") as f:
            self.labels = pickle.load(f)

    def __len__(self):
        return len(self.files)

    def __getitem__(self, idx):
        fp   = self.files[idx]
        x_td = np.load(fp)                        # (L, SIGLEN)
        # 频域幅度谱 (对称截取同长)
        x_fd = np.abs(np.fft.fft(x_td, axis=1)).astype("float32")
        x    = np.concatenate([x_td, x_fd], axis=0)  # (L*2, SIGLEN)
        y    = self.labels[fp.stem]
        return torch.tensor(x), torch.tensor(y, dtype=torch.long)


def load_data_ecg(batch_size: int = 256):
    train_set  = ECGDataset(True)
    test_set   = ECGDataset(False)
    classes    = len(set(train_set.labels.values()))
    g = torch.Generator()
    g.manual_seed(SEED)
    train_iter = data.DataLoader(train_set, batch_size, shuffle=True,  num_workers=0,
                                 drop_last=True, worker_init_fn=seed_worker, generator=g)
    test_iter  = data.DataLoader(test_set,  batch_size, shuffle=False, num_workers=0,
                                 worker_init_fn=seed_worker, generator=g)
    return train_iter, test_iter, classes

# ===================== 4. 模型 ===================== #
class ECGNet(nn.Module):
    def __init__(self, in_channels: int = IN_CHANNELS, classes: int = 5, drop_p: float = 0.3):
        super().__init__()
        self.net = nn.Sequential(
            nn.Conv1d(in_channels,  32, 7, 2, 3), nn.BatchNorm1d(32),  nn.ReLU(), nn.Dropout(drop_p),
            nn.Conv1d(32,  64, 5, 2, 2),          nn.BatchNorm1d(64),  nn.ReLU(), nn.Dropout(drop_p),
            nn.Conv1d(64, 128, 3, 2, 1),          nn.BatchNorm1d(128), nn.ReLU(), nn.Dropout(drop_p),
            nn.AdaptiveAvgPool1d(1), nn.Flatten(),
            nn.Linear(128, classes), nn.Dropout(drop_p)
        )

    def forward(self, x):
        return self.net(x)  # x: (B, C, T)


def init_weights(m):
    if isinstance(m, (nn.Conv1d, nn.Linear)):
        nn.init.kaiming_normal_(m.weight, nonlinearity="relu")

# ===================== 5. Early-Stopping ===================== #
class EarlyStopping:
    """在验证集 metric 连续 patience 轮不提升时停止训练"""
    def __init__(self, patience: int = 15, verbose: bool = True):
        self.patience = patience
        self.best_acc = 0.0
        self.num_bad  = 0
        self.verbose  = verbose
        self.best_path = "best_model.pth"

    def __call__(self, acc: float, model: nn.Module):
        if acc > self.best_acc:
            self.best_acc = acc
            self.num_bad  = 0
            torch.save(model.state_dict(), self.best_path)
            if self.verbose:
                print(f"  ✔ new best test_acc={acc:.3f} — model saved → {self.best_path}")
        else:
            self.num_bad += 1
            if self.verbose:
                print(f"  ✘ test_acc did not improve for {self.num_bad}/{self.patience} epochs")
        return self.num_bad >= self.patience

# ===================== 6. 训练 ===================== #

def accuracy(y_hat, y):
    return (y_hat.argmax(1) == y).sum()


def train(num_epochs: int = 30, batch: int = 256, lr: float = 1e-3, device: str | None = None,
          drop_p: float = 0.3, weight_decay: float = 1e-4, patience_lr: int = 2, patience_es: int = 15):
    device = device or (
        "mps"  if torch.backends.mps.is_available() else
        "cuda" if torch.cuda.is_available() else
        "cpu")

    train_iter, test_iter, C = load_data_ecg(batch)
    net = ECGNet(IN_CHANNELS, C, drop_p).to(device)
    net.apply(init_weights)

    loss_fn  = nn.CrossEntropyLoss()
    opt      = torch.optim.Adam(net.parameters(), lr, weight_decay=weight_decay)
    scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(opt, mode="max", factor=0.5,
                                                           patience=patience_lr, verbose=True)
    #stopper   = EarlyStopping(patience_es)
    animator  = d2l.Animator(xlabel="epoch", ylabel="acc", xlim=[1, num_epochs], legend=["train loss", "train acc", "test acc"])
    timer     = d2l.Timer()

    for epoch in range(num_epochs):
        metric = d2l.Accumulator(3)
        net.train()
        for X, y in train_iter:
            X, y = X.to(device, non_blocking=True), y.to(device, non_blocking=True)
            y_hat = net(X)
            l = loss_fn(y_hat, y)
            opt.zero_grad(); l.backward(); opt.step()
            metric.add(l.item() * y.numel(), accuracy(y_hat, y).item(), y.numel())

        train_loss, train_acc = metric[0] / metric[2], metric[1] / metric[2]
        net.eval(); test_metric = d2l.Accumulator(2)
        with torch.no_grad():
            for X, y in test_iter:
                X, y = X.to(device, non_blocking=True), y.to(device, non_blocking=True)
                test_metric.add(accuracy(net(X), y).item(), y.numel())
        test_acc = test_metric[0] / test_metric[1]
        scheduler.step(test_acc)

        animator.add(epoch + 1, (train_loss, train_acc, test_acc))
        display.clear_output(wait=True)
        display.display(animator.fig)
        animator.fig.savefig("training_curve.png", dpi=300, bbox_inches='tight')
        print(f"epoch {epoch + 1:2d}: loss {train_loss:.4f}  train_acc {train_acc:.3f}  test_acc {test_acc:.3f}")

        # Early-Stopping
      #  if stopper(test_acc, net):
       #     print("▲ Early stopping triggered. Stopping training.")
        #    break
    animator.axes[0].set_title("Lightweight 1D CNN Model Result")
    print(f"⏱  {timer.stop():.1f}s on {device}")
    print(f"✔ best test_acc={stopper.best_acc:.3f} — model weights saved to {stopper.best_path}")
# ===================== confusion matrix ===================== #
    print("\n▶ Generating confusion matrix for best model...")
    net.load_state_dict(torch.load(stopper.best_path, map_location=device))
    net.eval()
    y_true, y_pred = [], []
    with torch.no_grad():
        for X, y in test_iter:
            X, y = X.to(device), y.to(device)
            outputs = net(X)
            preds = outputs.argmax(1)
            y_true.extend(y.cpu().numpy())
            y_pred.extend(preds.cpu().numpy())

    cm = confusion_matrix(y_true, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["N", "V", "S", "F", "Q"])
    fig, ax = plt.subplots(figsize=(6, 6))
    disp.plot(ax=ax, cmap=plt.cm.Blues, values_format='d')
    plt.title("Confusion Matrix - ECG Classification")
    plt.savefig("confusion_matrix.png", dpi=300, bbox_inches='tight')
    plt.show()
# ===================== 主流程 ===================== #
if __name__ == "__main__":
    warnings.filterwarnings("ignore")
    download_mitbih()
    cut_beats()
    train()
```

```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-
"""
ECG Arrhythmia Classification — MIT-BIH → Multi-lead + FFT + Down-sampled 1-D CNN (PyTorch)
---------------------------------------------------------------------

  • Dropout 正则化（每个 Conv Block + 最后全连接）
  • Adam 权重衰减 (L2) = 1e-4
  • ReduceLROnPlateau 学习率调度 (factor 0.5, patience 2)
  • Early-Stopping (patience 5) — 自动保存最佳模型 best_model.pth
  • 固定随机种子保证可复现

用法：python ecg_arrhythmia_cnn_enhanced.py  # 直接跑完整流程
"""

import pickle, warnings, math, random, time
from pathlib import Path

import wfdb, numpy as np, neurokit2 as nk
import torch
from torch import nn
from torch.utils import data
from IPython import display  
from sklearn.preprocessing import StandardScaler
from tqdm import tqdm
from d2l import torch as d2l
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
import csv

# ================= 全局绘图样式 ================= #
plt.rcParams['font.family'] = 'Arial'   # 字体
plt.rcParams['font.size'] = 11          # 字号
plt.rcParams['axes.titlesize'] = 11
plt.rcParams['axes.labelsize'] = 11
plt.rcParams['xtick.labelsize'] = 11
plt.rcParams['ytick.labelsize'] = 11
plt.rcParams['legend.fontsize'] = 11

# ===================== 固定随机种子 ===================== #
SEED = 42
random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(SEED)

def seed_worker(_):
    np.random.seed(SEED)
    random.seed(SEED)

# ===================== 常量 ===================== #
ROOT        = Path("/Users/gaochenxi/datasets/mitbih")
RAW_DIR     = ROOT                      # wfdb 下载到此
MITDB_DIR   = RAW_DIR                   # 48 条记录目录
TRAIN_DIR   = ROOT / "train"            # 训练拍 .npy
TEST_DIR    = ROOT / "test"             # 测试拍 .npy
LABELS_PKL  = ROOT / "labels.pkl"

# 信号相关
LEADS       = [0, 1]      # MLII, V5
ORIG_FS     = 360         # MIT-BIH 原采样率
TARGET_FS   = 180         # 降采样目标采样率 (ORIG_FS // 2)
SIGLEN      = 128         # 降采样后固定长度 (samples)
LEFT_PAD    = 40          # R 峰距窗口左端 (TARGET_FS 下)

TEST_RECS = {str(r) for r in [101,106,108,109,112,114,115,116,118,119,
                              122,124,201,203,205,207,208,209,215,220,223,230,234]}
MIT2AAMI  = {'N':0,'L':0,'R':0,'e':0,'j':0,
             'V':1,'E':1,
             'A':2,'a':2,'J':2,'S':2,
             'F':3,
             'Q':4,'|':4,'f':4}


LEAD_N        = len(LEADS)          # 2
IN_CHANNELS   = LEAD_N * 2          # 时域 + 频域
AUG_FACTOR    = ORIG_FS / TARGET_FS # 2.0
LEFT_PAD_RAW  = int(LEFT_PAD * AUG_FACTOR)
SIGLEN_RAW    = int(SIGLEN * AUG_FACTOR)

# ===================== 1. 下载 ===================== #

def download_mitbih() -> None:
    """确保 MIT-BIH 48 条记录在本地"""
    need_redownload = (
        not MITDB_DIR.exists() or
        len(list(MITDB_DIR.glob("*.dat"))) < 48 or
        len(list(MITDB_DIR.glob("*.hea"))) < 48 or
        len(list(MITDB_DIR.glob("*.atr"))) < 48
    )
    if not need_redownload:
        print("✔ MIT-BIH 已完整下载，跳过")
        return

    print("▶ (重新) 下载 MIT-BIH →", RAW_DIR)
    import shutil
    shutil.rmtree(MITDB_DIR.parent, ignore_errors=True)
    RAW_DIR.mkdir(parents=True, exist_ok=True)
    wfdb.dl_database("mitdb", dl_dir=str(RAW_DIR))  # 自动生成 mitdb/1.0.0
    print("✔ 下载完成")

# ===================== 2. 切单拍 ===================== #

def _filter(sig: np.ndarray, fs: int = ORIG_FS) -> np.ndarray:
    """0.5–40 Hz 带通滤波 (neurokit2)"""
    return nk.signal_filter(sig, sampling_rate=fs, lowcut=0.5, highcut=40)


def _resample(sig: np.ndarray, desired_len: int = SIGLEN) -> np.ndarray:
    """重采样至 desired_len 样本 (≈ TARGET_FS)"""
    return nk.signal_resample(sig, desired_length=desired_len)


def cut_beats() -> None:
    """把 .dat 切成多导联单拍 .npy，并做 Z-score 标准化 (train+test)"""
    if TRAIN_DIR.exists() and any(TRAIN_DIR.glob("*.npy")):
        print("✔ 已有 .npy，跳过 cut_beats()")
        return

    TRAIN_DIR.mkdir(parents=True, exist_ok=True)
    TEST_DIR.mkdir(parents=True,  exist_ok=True)

    hea_files = sorted(MITDB_DIR.glob("*.hea"))
    if len(hea_files) != 48:
        raise FileNotFoundError("MIT-BIH 记录不全，请检查下载目录")

    scaler = StandardScaler()
    print("▶ Pass-1 计算均值/方差 (全部导联)")
    for hea in tqdm(hea_files):
        sig, _ = wfdb.rdsamp(str(hea.with_suffix("")), channels=LEADS)
        sig = np.vstack([_filter(sig[:, i]) for i in range(LEAD_N)]).T  # shape (T, L)
        scaler.partial_fit(sig.reshape(-1, 1))

    print("▶ Pass-2 切单拍并保存 .npy")
    labels = {}
    for hea in tqdm(hea_files):
        rec = hea.stem
        record = wfdb.rdrecord(str(hea.with_suffix("")), channels=LEADS)
        ann    = wfdb.rdann  (str(hea.with_suffix("")), "atr")

        sig = record.p_signal.astype("float32")  # shape (T, L)
        sig = np.vstack([_filter(sig[:, i]) for i in range(LEAD_N)]).T
        sig = scaler.transform(sig.reshape(-1, 1)).reshape(-1, LEAD_N)  # Z-score

        for idx, (r, sym) in enumerate(zip(ann.sample, ann.symbol)):
            if sym not in MIT2AAMI:
                continue
            start, end = r - LEFT_PAD_RAW, r - LEFT_PAD_RAW + SIGLEN_RAW
            if start < 0 or end > len(sig):
                continue
            beat = sig[start:end]             # shape (SIGLEN_RAW, L)
            beat = beat.T                     # (L, SIGLEN_RAW)
            beat = np.stack([_resample(ch, SIGLEN) for ch in beat])  # (L, SIGLEN)
            fname   = f"{rec}_{idx:05d}.npy"
            target  = TEST_DIR if rec in TEST_RECS else TRAIN_DIR
            np.save(target / fname, beat.astype("float32"))
            labels[f"{rec}_{idx:05d}"] = MIT2AAMI[sym]

    with open(LABELS_PKL, "wb") as f:
        pickle.dump(labels, f)

    print(f"✔ 保存单拍: train={len(list(TRAIN_DIR.glob('*.npy')))}, "
          f"test={len(list(TEST_DIR.glob('*.npy')))}")

# ===================== 3. Dataset ===================== #
class ECGDataset(data.Dataset):
    """按需拼接时域 + FFT 通道"""
    def __init__(self, train: bool = True):
        self.dir   = TRAIN_DIR if train else TEST_DIR
        self.files = sorted(self.dir.glob("*.npy"))
        with open(LABELS_PKL, "rb") as f:
            self.labels = pickle.load(f)

    def __len__(self):
        return len(self.files)

    def __getitem__(self, idx):
        fp   = self.files[idx]
        x_td = np.load(fp)                        # (L, SIGLEN)
        # 频域幅度谱 (对称截取同长)
        x_fd = np.abs(np.fft.fft(x_td, axis=1)).astype("float32")
        x    = np.concatenate([x_td, x_fd], axis=0)  # (L*2, SIGLEN)
        y    = self.labels[fp.stem]
        return torch.tensor(x), torch.tensor(y, dtype=torch.long)


def load_data_ecg(batch_size: int = 256):
    train_set  = ECGDataset(True)
    test_set   = ECGDataset(False)
    classes    = len(set(train_set.labels.values()))
    g = torch.Generator()
    g.manual_seed(SEED)
    train_iter = data.DataLoader(train_set, batch_size, shuffle=True,  num_workers=0,
                                 drop_last=True, worker_init_fn=seed_worker, generator=g)
    test_iter  = data.DataLoader(test_set,  batch_size, shuffle=False, num_workers=0,
                                 worker_init_fn=seed_worker, generator=g)
    return train_iter, test_iter, classes

# ===================== 4. 模型 ===================== #
class ECGNet(nn.Module):
    def __init__(self, in_channels: int = IN_CHANNELS, classes: int = 5, drop_p: float = 0.3):
        super().__init__()
        self.net = nn.Sequential(
            nn.Conv1d(in_channels,  32, 7, 2, 3), nn.BatchNorm1d(32),  nn.ReLU(), nn.Dropout(drop_p),
            nn.Conv1d(32,  64, 5, 2, 2),          nn.BatchNorm1d(64),  nn.ReLU(), nn.Dropout(drop_p),
            nn.Conv1d(64, 128, 3, 2, 1),          nn.BatchNorm1d(128), nn.ReLU(), nn.Dropout(drop_p),
            nn.AdaptiveAvgPool1d(1), nn.Flatten(),
            nn.Linear(128, classes), nn.Dropout(drop_p)
        )

    def forward(self, x):
        return self.net(x)  # x: (B, C, T)


def init_weights(m):
    if isinstance(m, (nn.Conv1d, nn.Linear)):
        nn.init.kaiming_normal_(m.weight, nonlinearity="relu")

# ===================== 5. Early-Stopping ===================== #
class EarlyStopping:
    """在验证集 metric 连续 patience 轮不提升时停止训练"""
    def __init__(self, patience: int = 10, verbose: bool = True):
        self.patience = patience
        self.best_acc = 0.0
        self.num_bad  = 0
        self.verbose  = verbose
        self.best_path = "best_model.pth"

    def __call__(self, acc: float, model: nn.Module):
        if acc > self.best_acc:
            self.best_acc = acc
            self.num_bad  = 0
            torch.save(model.state_dict(), self.best_path)
            if self.verbose:
                print(f"  ✔ new best test_acc={acc:.3f} — model saved → {self.best_path}")
        else:
            self.num_bad += 1
            if self.verbose:
                print(f"  ✘ test_acc did not improve for {self.num_bad}/{self.patience} epochs")
        return self.num_bad >= self.patience

# ===================== 6. 训练 ===================== #

def accuracy(y_hat, y):
    return (y_hat.argmax(1) == y).sum()


def train(num_epochs: int = 15, batch: int = 1024, lr: float = 1e-3, device: str | None = None,
          drop_p: float = 0.3, weight_decay: float = 1e-4, patience_lr: int = 2, patience_es: int = 10):
    device = device or (
        "mps"  if torch.backends.mps.is_available() else
        "cuda" if torch.cuda.is_available() else
        "cpu")

    train_iter, test_iter, C = load_data_ecg(batch)
    net = ECGNet(IN_CHANNELS, C, drop_p).to(device)
    net.apply(init_weights)

    loss_fn  = nn.CrossEntropyLoss()
    opt      = torch.optim.Adam(net.parameters(), lr, weight_decay=weight_decay)
    scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(opt, mode="max", factor=0.5,
                                                           patience=patience_lr, verbose=True)
    stopper   = EarlyStopping(patience_es)
    animator  = d2l.Animator(xlabel="epoch", ylabel="acc", xlim=[1, num_epochs], legend=["train loss", "train acc", "test acc"])
    timer     = d2l.Timer()
    
    epochs_list, train_losses, train_accs, test_accs = [], [], [], []
    
    for epoch in range(num_epochs):
        metric = d2l.Accumulator(3)
        net.train()
        for X, y in train_iter:
            X, y = X.to(device, non_blocking=True), y.to(device, non_blocking=True)
            y_hat = net(X)
            l = loss_fn(y_hat, y)
            opt.zero_grad(); l.backward(); opt.step()
            metric.add(l.item() * y.numel(), accuracy(y_hat, y).item(), y.numel())

        train_loss, train_acc = metric[0] / metric[2], metric[1] / metric[2]
        net.eval(); test_metric = d2l.Accumulator(2)
        with torch.no_grad():
            for X, y in test_iter:
                X, y = X.to(device, non_blocking=True), y.to(device, non_blocking=True)
                test_metric.add(accuracy(net(X), y).item(), y.numel())
        test_acc = test_metric[0] / test_metric[1]
        scheduler.step(test_acc)
        #=== record to list==
        epochs_list.append(epoch + 1)
        train_losses.append(train_loss)
        train_accs.append(train_acc)
        test_accs.append(test_acc)
        
        animator.add(epoch + 1, (train_loss, train_acc, test_acc))
        display.clear_output(wait=True)
        display.display(animator.fig)
        animator.fig.savefig("training_curve.png", dpi=300, bbox_inches='tight')
        print(f"epoch {epoch + 1:2d}: loss {train_loss:.4f}  train_acc {train_acc:.3f}  test_acc {test_acc:.3f}")

        # Early-Stopping
        if stopper(test_acc, net):
            print("▲ Early stopping triggered. Stopping training.")
            break
    animator.axes[0].set_title("Lightweight 1D CNN Model Result")
    print(f"⏱  {timer.stop():.1f}s on {device}")
    print(f"✔ best test_acc={stopper.best_acc:.3f} — model weights saved to {stopper.best_path}")

# ===== 新增：保存 CSV 日志 =====
    with open("training_log.csv", "w", newline="") as f:
        w = csv.writer(f)
        w.writerow(["epoch", "train_loss", "train_acc", "test_acc"])
        for e, tl, ta, te in zip(epochs_list, train_losses, train_accs, test_accs):
            w.writerow([e, tl, ta, te])
    print("✔ saved metrics → training_log.csv")

 # ===== 新增：绘制完整曲线并保存 =====
    plt.figure(figsize=(6.5, 4.2))
    plt.plot(epochs_list, train_losses, label="train loss")
    plt.plot(epochs_list, train_accs,  label="train acc")
    plt.plot(epochs_list, test_accs,   label="test acc")
    plt.xlabel("Epoch")
    plt.ylabel("Metric")
    plt.title("Training Curve")
    plt.legend()
    plt.grid(True, linestyle="--", alpha=0.5)
    plt.tight_layout()
    plt.savefig("training_curve.png", dpi=300, bbox_inches="tight")
    plt.show()
    print("✔ saved figure → training_curve.png")
# ===================== confusion matrix ===================== #
    print("\n▶ Generating confusion matrix for best model...")
    net.load_state_dict(torch.load(stopper.best_path, map_location=device))
    net.eval()
    y_true, y_pred = [], []
    with torch.no_grad():
        for X, y in test_iter:
            X, y = X.to(device), y.to(device)
            outputs = net(X)
            preds = outputs.argmax(1)
            y_true.extend(y.cpu().numpy())
            y_pred.extend(preds.cpu().numpy())

    cm = confusion_matrix(y_true, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["N", "V", "S", "F", "Q"])
    fig, ax = plt.subplots(figsize=(6, 6))
    disp.plot(ax=ax, cmap=plt.cm.Blues, values_format='d')
    plt.title("Confusion Matrix - ECG Classification")
    plt.savefig("confusion_matrix.png", dpi=300, bbox_inches='tight')
    plt.show()
# ===================== 主流程 ===================== #
if __name__ == "__main__":
    warnings.filterwarnings("ignore")
    download_mitbih()
    cut_beats()
    train()
```

```python
# ==========================================================
# 4-th Order Butterworth Band-Pass (0.5–40 Hz)  ▶  Bode Plot
# ==========================================================
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import butter, freqz
import matplotlib.pyplot as plt

# ================= 全局绘图样式 ================= #
plt.rcParams['font.family'] = 'Arial'   # 字体
plt.rcParams['font.size'] = 11          # 字号
plt.rcParams['axes.titlesize'] = 11
plt.rcParams['axes.labelsize'] = 11
plt.rcParams['xtick.labelsize'] = 11
plt.rcParams['ytick.labelsize'] = 11
plt.rcParams['legend.fontsize'] = 11

# ==== 可调参数 =============================================
fs      = 360.0          # 采样率 (Hz)
lowcut  = 0.5            # 下截止 (Hz)
highcut = 40.0           # 上截止 (Hz)
order   = 4              # Butterworth 阶数
n_fft   = 4096           # 频率采样点数（越大曲线越平滑）
# ==========================================================

# 1. 设计数字 Butterworth 带通（SciPy 直接用双线性）
b, a = butter(order, Wn=[lowcut, highcut], btype='bandpass', fs=fs)

# 2. 计算频率响应
w, h = freqz(b, a, worN=n_fft, fs=fs)   # w: 0..fs/2

# 3. 绘图
plt.figure(figsize=(8, 4))
plt.plot(w, np.abs(h), linewidth=2)
plt.title(f'{order}th-Order Butterworth Band Pass Filter ({lowcut}–{highcut} Hz, fs={fs} Hz)')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude |H(f)|')
plt.ylim(-0.05, 1.05)
plt.xlim(0, fs/2)
plt.grid(True, linestyle='--', alpha=0.6)
plt.savefig("Butterworth Band Pass filter.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import neurokit2 as nk

# 生成模拟 ECG 信号（含噪声）
fs = 360  # 采样率
duration = 2  # 秒
ecg = nk.ecg_simulate(duration=duration, sampling_rate=fs)

# 添加低频漂移 & 高频噪声
t = np.arange(len(ecg)) / fs
low_freq_drift = 0.2 * np.sin(2 * np.pi * 0.2 * t)
high_freq_noise = 0.05 * np.sin(2 * np.pi * 60 * t)
noisy_ecg = ecg + low_freq_drift + high_freq_noise

# 带通滤波 0.5–40 Hz
filtered_ecg = nk.signal_filter(noisy_ecg, sampling_rate=fs, lowcut=0.5, highcut=40)

# 绘图
plt.figure(figsize=(10, 4))
plt.plot(t, noisy_ecg, label="Original with drift + noise", alpha=0.6)
plt.plot(t, filtered_ecg, label="After 0.5–40 Hz Bandpass", linewidth=2)
plt.xlabel("Time (s)")
plt.ylabel("Amplitude (mV)")
plt.title("Effect of 0.5–40 Hz Bandpass Filter on ECG Signal")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# ---------- timeline ----------
t_end = 0.8                 # 秒
fs = 2000                   # 为了曲线平滑，仅用于绘图
t = np.linspace(0, t_end, int(t_end*fs), endpoint=False)

# ---------- 构造 ECG 类波形 ----------
y = np.zeros_like(t)

# P 波（0.10s 左右的小包）
y += 0.12 * np.exp(-0.5 * ((t - 0.10) / 0.025) ** 2)

# PR segment（略高于基线）
y += ((t > 0.12) & (t < 0.18)) * 0.02

# QRS 复合波（在 0.22s 附近的尖峰：Q 下降、R 上冲、S 下冲，用三角近似）
def qrs_triangle(center, left_w, right_w, q_drop=-0.25, r_height=1.05, s_drop=-0.45):
    yq = np.zeros_like(t)
    q_start = center - left_w
    r_time  = center
    s_end   = center + right_w
    # Q -> R
    idx_q = (t >= q_start) & (t < r_time)
    yq[idx_q] = np.interp(t[idx_q], [q_start, r_time], [q_drop, r_height])
    # R -> S
    idx_s = (t >= r_time) & (t <= s_end)
    yq[idx_s] = np.interp(t[idx_s], [r_time, s_end], [r_height, s_drop])
    return yq

y += qrs_triangle(0.22, left_w=0.018, right_w=0.030)

# ST segment（轻度抬高的平段）
y += ((t > 0.27) & (t < 0.40)) * 0.03

# T 波（0.58s 的圆顶）
y += 0.45 * np.exp(-0.5 * ((t - 0.58) / 0.045) ** 2)

# ---------- ECG 方格纸网格 ----------
fig, ax = plt.subplots(figsize=(10, 5))
ax.set_xlim(0, t_end)
ax.set_ylim(-0.8, 1.2)

# 模拟 1mm/5mm：时间 0.04s/0.2s；幅度 0.1/0.5（相对单位）
x_minor, x_major = 0.04, 0.20
y_minor, y_major = 0.10, 0.50

ax.set_xticks(np.arange(0, t_end + 1e-9, x_major))
ax.set_xticks(np.arange(0, t_end + 1e-9, x_minor), minor=True)
ax.set_yticks(np.arange(-2, 2, y_major))
ax.set_yticks(np.arange(-2, 2, y_minor), minor=True)
ax.grid(which='major', linewidth=1.0, alpha=0.35)
ax.grid(which='minor', linewidth=0.5, alpha=0.25)

# ---------- 绘制波形 ----------
ax.plot(t, y, linewidth=3, color='black')

# ---------- 常用标注 ----------
def y_at(x):  # 取曲线在某个时间点的值
    return y[np.argmin(np.abs(t - x))]

# 标注 P、Q、R、S、T
ax.text(0.10, y_at(0.10)+0.08, "P", ha='center', va='bottom', fontsize=12, fontweight='bold')
ax.text(0.205, y_at(0.205)-0.12, "Q", ha='center', va='top', fontsize=12, fontweight='bold')
ax.text(0.220, y_at(0.220)+0.05, "R", ha='center', va='bottom', fontsize=12, fontweight='bold')
ax.text(0.245, y_at(0.245)-0.12, "S", ha='center', va='top', fontsize=12, fontweight='bold')
ax.text(0.58,  y_at(0.58)+0.05,  "T", ha='center', va='bottom', fontsize=12, fontweight='bold')

# R-peak 圈圈
ax.scatter([0.22], [y_at(0.22)], s=60, facecolors='none', edgecolors='k')
ax.text(0.225, y_at(0.22)+0.10, "R-peak", va='bottom')

# PR segment
ax.annotate("", xy=(0.12, 0.10), xytext=(0.18, 0.10),
            arrowprops=dict(arrowstyle="<->", lw=1.2))
ax.text(0.15, 0.12, "PR segment", ha='center', va='bottom')

# ST segment
ax.annotate("", xy=(0.30, 0.06), xytext=(0.40, 0.06),
            arrowprops=dict(arrowstyle="<->", lw=1.2))
ax.text(0.35, 0.08, "ST segment", ha='center', va='bottom')

# PR interval（0 → Q 起点 ~0.205s）
ax.annotate("", xy=(0.00, -0.65), xytext=(0.205, -0.65),
            arrowprops=dict(arrowstyle="<->", lw=1.2))
ax.text(0.102, -0.62, "PR interval", ha='center', va='bottom')

# QT interval（Q 起点 ~0.205s → T 末 ~0.66s）
ax.annotate("", xy=(0.205, -0.72), xytext=(0.66, -0.72),
            arrowprops=dict(arrowstyle="<->", lw=1.2))
ax.text(0.432, -0.69, "QT interval", ha='center', va='bottom')

# 文字说明（大致位置）
ax.text(0.27, 0.65, "QRS complex", ha='center', va='bottom')
ax.text(0.30, 0.32, "Ventricular\nDepolarization", ha='left', va='center')
ax.text(0.57, 0.45, "Ventricular\nRepolarization", ha='center', va='center')
ax.text(0.06, 0.22, "Atrial\nDepolarization", ha='center', va='center')

ax.set_xlabel("Time (s)")
ax.set_ylabel("Amplitude (a.u.)")
ax.set_title("ECG schematic with PR/QRS/ST/T and intervals")
plt.tight_layout()
plt.show()
```

```python
# single_beat_plot.py
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
import pickle
import random

# ====== 路径 & 常量（按你的工程保持一致）======
ROOT        = Path("/Users/gaochenxi/datasets/mitbih")
TRAIN_DIR   = ROOT / "train"
TEST_DIR    = ROOT / "test"
LABELS_PKL  = ROOT / "labels.pkl"

SIGLEN      = 256        # 降采样后固定长度
TARGET_FS   = 180        # 用于横轴单位换算（秒）
LEFT_PAD    = 80         # R 峰左侧样本点数（在 TARGET_FS 下）
LEAD_NAMES  = ["MLII", "V5"]
ID2NAME     = {0:"N", 1:"V", 2:"S", 3:"F", 4:"Q"}

# ====== 全局绘图样式（Arial 11，导出清晰）======
plt.rcParams['font.family'] = 'Arial'
plt.rcParams['font.size']   = 11
plt.rcParams['axes.titlesize'] = 11
plt.rcParams['axes.labelsize'] = 11
plt.rcParams['xtick.labelsize'] = 11
plt.rcParams['ytick.labelsize'] = 11
plt.rcParams['legend.fontsize'] = 11

def load_labels():
    with open(LABELS_PKL, "rb") as f:
        return pickle.load(f)  # dict: {stem -> class_id}

def load_beat_npy(npy_path: Path) -> np.ndarray:
    """读取单拍 .npy，返回时域 2×256（若是 4×256，取前两行）"""
    x = np.load(npy_path)           # (2,256) 或 (4,256)
    if x.shape[0] == 4:
        x = x[:2]                   # 只取时域 MLII, V1
    assert x.shape == (2, SIGLEN), f"unexpected shape: {x.shape}"
    return x.astype("float32")

def plot_single_beat(npy_path: Path, save_path: Path | None = None,
                     title_prefix: str = "Single Beat"):
    """画单拍：两导联各一行；横轴以 R 峰为 0 s"""
    labels = load_labels()
    stem   = npy_path.stem
    y_id   = labels.get(stem, None)
    y_name = ID2NAME.get(y_id, "?")

    x = load_beat_npy(npy_path)     # (2,256)
    # 构造相对 R 峰的时间轴（秒）：[-LEFT/FS, (SIGLEN-LEFT)/FS)
    t = (np.arange(SIGLEN) - LEFT_PAD) / TARGET_FS

    fig, axes = plt.subplots(2, 1, figsize=(7.2, 4.8), sharex=True)
    for i, ax in enumerate(axes):
        ax.plot(t, x[i], linewidth=1.8)
        ax.axvline(0.0, color='k', linestyle='--', linewidth=1, alpha=0.7)  # R 处竖线
        ax.set_ylabel(f"{LEAD_NAMES[i]} (z-score)")
        ax.grid(True, linestyle='--', alpha=0.35)

    axes[-1].set_xlabel("Time (s)")
    fig.suptitle(f"{title_prefix} — {stem}  (Label: {y_name})")

    plt.tight_layout(rect=[0, 0, 1, 0.96])

    if save_path is None:
        save_path = npy_path.with_suffix(".png")
    fig.savefig(save_path, dpi=300, bbox_inches='tight')
    plt.show()
    print(f"✔ saved → {save_path}")

# ====== 示例用法 ======
if __name__ == "__main__":
    # 任选一个样本（也可以手动指定文件名）
    candidates = sorted(list(TRAIN_DIR.glob("*.npy"))) + sorted(list(TEST_DIR.glob("*.npy")))
    assert candidates, "未找到 .npy 单拍文件，请先运行切拍流程。"
    npy = random.choice(candidates)
    plot_single_beat(npy, save_path=ROOT / "single_beat_example.png")
```

```python
import pickle, warnings, math, random, time
from pathlib import Path

import wfdb, numpy as np, neurokit2 as nk
import torch
from torch import nn
from torch.utils import data
from IPython import display  
from sklearn.preprocessing import StandardScaler
from tqdm import tqdm
from d2l import torch as d2l
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
import csv

# ================= GLobal plotting style ================= #
plt.rcParams['font.family'] = 'Arial'   
plt.rcParams['font.size'] = 11          
plt.rcParams['axes.titlesize'] = 11
plt.rcParams['axes.labelsize'] = 11
plt.rcParams['xtick.labelsize'] = 11
plt.rcParams['ytick.labelsize'] = 11
plt.rcParams['legend.fontsize'] = 11

# ===================== Random Seed ===================== #
SEED = 42
random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(SEED)

def seed_worker(_):
    np.random.seed(SEED)
    random.seed(SEED)

# ===================== Constants ===================== #
ROOT        = Path("/Users/gaochenxi/datasets/mitbih")
RAW_DIR     = ROOT                      # wfdb will download here
MITDB_DIR   = RAW_DIR                   # 48 records
TRAIN_DIR   = ROOT / "train"            # training beats .npy
TEST_DIR    = ROOT / "test"             # testing beats .npy
LABELS_PKL  = ROOT / "labels.pkl"

# 信号相关
LEADS       = [0, 1]      # MLII, V5
ORIG_FS     = 360         # MIT-BIH original sampling rate
TARGET_FS   = 180         # Downsample target sampling rate (ORIG_FS // 2)
SIGLEN      = 128         # Fixed length after downsampling (samples)
LEFT_PAD    = 40          # Left padding for R peak window (in TARGET_FS)

TEST_RECS = {str(r) for r in [101,106,108,109,112,114,115,116,118,119,
                              122,124,201,203,205,207,208,209,215,220,223,230,234]}
MIT2AAMI  = {'N':0,'L':0,'R':0,'e':0,'j':0,
             'V':1,'E':1,
             'A':2,'a':2,'J':2,'S':2,
             'F':3,
             'Q':4,'|':4,'f':4}


LEAD_N        = len(LEADS)          # 2
IN_CHANNELS   = LEAD_N * 2          # time domain + frequency domain
AUG_FACTOR    = ORIG_FS / TARGET_FS # 2.0
LEFT_PAD_RAW  = int(LEFT_PAD * AUG_FACTOR)
SIGLEN_RAW    = int(SIGLEN * AUG_FACTOR)

# ===================== 1. Download ===================== #

def download_mitbih() -> None:
    """Ensure MIT-BIH 48 records exist locally"""
    need_redownload = (
        not MITDB_DIR.exists() or
        len(list(MITDB_DIR.glob("*.dat"))) < 48 or
        len(list(MITDB_DIR.glob("*.hea"))) < 48 or
        len(list(MITDB_DIR.glob("*.atr"))) < 48
    )
    if not need_redownload:
        print("✔ MIT-BIH already downloaded, skipping")
        return

    print("▶ (Re) Downloading MIT-BIH →", RAW_DIR)
    import shutil
    shutil.rmtree(MITDB_DIR.parent, ignore_errors=True)
    RAW_DIR.mkdir(parents=True, exist_ok=True)
    wfdb.dl_database("mitdb", dl_dir=str(RAW_DIR))  # Auto creat mitdb/1.0.0
    print("✔ Download complete")

# ===================== 2. Beat Segmentation ===================== #

def _filter(sig: np.ndarray, fs: int = ORIG_FS) -> np.ndarray:
    """0.5–40 Hz bandpass filter (neurokit2)"""
    return nk.signal_filter(sig, sampling_rate=fs, lowcut=0.5, highcut=40)


def _resample(sig: np.ndarray, desired_len: int = SIGLEN) -> np.ndarray:
    """Resample signal to desired_len samples (≈ TARGET_FS)"""
    return nk.signal_resample(sig, desired_length=desired_len)


def cut_beats() -> None:
    """Split .dat files into multi-lead beats .npy，and apply Z-score normalization (train+test)"""
    if TRAIN_DIR.exists() and any(TRAIN_DIR.glob("*.npy")):
        print("✔ .npy，files already exist, skipping cut_beats()")
        return

    TRAIN_DIR.mkdir(parents=True, exist_ok=True)
    TEST_DIR.mkdir(parents=True,  exist_ok=True)

    hea_files = sorted(MITDB_DIR.glob("*.hea"))
    if len(hea_files) != 48:
        raise FileNotFoundError("MIT-BIH records incomplete，please check download directory")

    scaler = StandardScaler()
    print("▶ Pass-1 compute mean/std (all leads)")
    for hea in tqdm(hea_files):
        sig, _ = wfdb.rdsamp(str(hea.with_suffix("")), channels=LEADS)
        sig = np.vstack([_filter(sig[:, i]) for i in range(LEAD_N)]).T  # shape (T, L)
        scaler.partial_fit(sig.reshape(-1, 1))

    print("▶ Pass-2 segment beat and save .npy")
    labels = {}
    for hea in tqdm(hea_files):
        rec = hea.stem
        record = wfdb.rdrecord(str(hea.with_suffix("")), channels=LEADS)
        ann    = wfdb.rdann  (str(hea.with_suffix("")), "atr")

        sig = record.p_signal.astype("float32")  # shape (T, L)
        sig = np.vstack([_filter(sig[:, i]) for i in range(LEAD_N)]).T
        sig = scaler.transform(sig.reshape(-1, 1)).reshape(-1, LEAD_N)  # Z-score

        for idx, (r, sym) in enumerate(zip(ann.sample, ann.symbol)):
            if sym not in MIT2AAMI:
                continue
            start, end = r - LEFT_PAD_RAW, r - LEFT_PAD_RAW + SIGLEN_RAW
            if start < 0 or end > len(sig):
                continue
            beat = sig[start:end]             # shape (SIGLEN_RAW, L)
            beat = beat.T                     # (L, SIGLEN_RAW)
            beat = np.stack([_resample(ch, SIGLEN) for ch in beat])  # (L, SIGLEN)
            fname   = f"{rec}_{idx:05d}.npy"
            target  = TEST_DIR if rec in TEST_RECS else TRAIN_DIR
            np.save(target / fname, beat.astype("float32"))
            labels[f"{rec}_{idx:05d}"] = MIT2AAMI[sym]

    with open(LABELS_PKL, "wb") as f:
        pickle.dump(labels, f)

    print(f"✔ Beats saved: train={len(list(TRAIN_DIR.glob('*.npy')))}, "
          f"test={len(list(TEST_DIR.glob('*.npy')))}")

# ===================== 3. Dataset ===================== #
class ECGDataset(data.Dataset):
    """Concatenate time domain + FFT channels"""
    def __init__(self, train: bool = True):
        self.dir   = TRAIN_DIR if train else TEST_DIR
        self.files = sorted(self.dir.glob("*.npy"))
        with open(LABELS_PKL, "rb") as f:
            self.labels = pickle.load(f)

    def __len__(self):
        return len(self.files)

    def __getitem__(self, idx):
        fp   = self.files[idx]
        x_td = np.load(fp)                        # (L, SIGLEN)
        # Magnitude spectrum in frequency domain with the same temporal length
        x_fd = np.abs(np.fft.fft(x_td, axis=1)).astype("float32")
        x    = np.concatenate([x_td, x_fd], axis=0)  # (L*2, SIGLEN)
        y    = self.labels[fp.stem]
        return torch.tensor(x), torch.tensor(y, dtype=torch.long)


def load_data_ecg(batch_size: int = 256):
    train_set  = ECGDataset(True)
    test_set   = ECGDataset(False)
    classes    = len(set(train_set.labels.values()))
    g = torch.Generator()
    g.manual_seed(SEED)
    train_iter = data.DataLoader(train_set, batch_size, shuffle=True,  num_workers=0,
                                 drop_last=True, worker_init_fn=seed_worker, generator=g)
    test_iter  = data.DataLoader(test_set,  batch_size, shuffle=False, num_workers=0,
                                 worker_init_fn=seed_worker, generator=g)
    return train_iter, test_iter, classes

# ===================== 4. model ===================== #
class ECGNet(nn.Module):
    def __init__(self, in_channels: int = IN_CHANNELS, classes: int = 5, drop_p: float = 0.3):
        super().__init__()
        self.net = nn.Sequential(
            nn.Conv1d(in_channels,  32, 7, 2, 3), nn.BatchNorm1d(32),  nn.ReLU(), nn.Dropout(drop_p),
            nn.Conv1d(32,  64, 5, 2, 2),          nn.BatchNorm1d(64),  nn.ReLU(), nn.Dropout(drop_p),
            nn.Conv1d(64, 128, 3, 2, 1),          nn.BatchNorm1d(128), nn.ReLU(), nn.Dropout(drop_p),
            nn.AdaptiveAvgPool1d(1), nn.Flatten(),
            nn.Linear(128, classes), nn.Dropout(drop_p)
        )

    def forward(self, x):
        return self.net(x)  # x: (B, C, T)


def init_weights(m):
    if isinstance(m, (nn.Conv1d, nn.Linear)):
        nn.init.kaiming_normal_(m.weight, nonlinearity="relu")

# ===================== 5. Early-Stopping ===================== #
class EarlyStopping:
    """Stop training when a  validation metric fails to improve for patience epochs"""
    def __init__(self, patience: int = 10, verbose: bool = True):
        self.patience = patience
        self.best_acc = 0.0
        self.num_bad  = 0
        self.verbose  = verbose
        self.best_path = "best_model.pth"

    def __call__(self, acc: float, model: nn.Module):
        if acc > self.best_acc:
            self.best_acc = acc
            self.num_bad  = 0
            torch.save(model.state_dict(), self.best_path)
            if self.verbose:
                print(f"  ✔ new best test_acc={acc:.3f} — model saved → {self.best_path}")
        else:
            self.num_bad += 1
            if self.verbose:
                print(f"  ✘ test_acc did not improve for {self.num_bad}/{self.patience} epochs")
        return self.num_bad >= self.patience

# ===================== 6. Train ===================== #

def accuracy(y_hat, y):
    return (y_hat.argmax(1) == y).sum()


def train(num_epochs: int = 15, batch: int = 1024, lr: float = 1e-3, device: str | None = None,
          drop_p: float = 0.3, weight_decay: float = 1e-4, patience_lr: int = 2, patience_es: int = 10):
    device = device or (
        "mps"  if torch.backends.mps.is_available() else
        "cuda" if torch.cuda.is_available() else
        "cpu")

    train_iter, test_iter, C = load_data_ecg(batch)
    net = ECGNet(IN_CHANNELS, C, drop_p).to(device)
    net.apply(init_weights)

    loss_fn  = nn.CrossEntropyLoss()
    opt      = torch.optim.Adam(net.parameters(), lr, weight_decay=weight_decay)
    scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(opt, mode="max", factor=0.5,
                                                           patience=patience_lr, verbose=True)
    stopper   = EarlyStopping(patience_es)
    animator  = d2l.Animator(xlabel="epoch", ylabel="acc", xlim=[1, num_epochs], legend=["train loss", "train acc", "test acc"])
    timer     = d2l.Timer()
    
    epochs_list, train_losses, train_accs, test_accs = [], [], [], []
    
    for epoch in range(num_epochs):
        metric = d2l.Accumulator(3)
        net.train()
        for X, y in train_iter:
            X, y = X.to(device, non_blocking=True), y.to(device, non_blocking=True)
            y_hat = net(X)
            l = loss_fn(y_hat, y)
            opt.zero_grad(); l.backward(); opt.step()
            metric.add(l.item() * y.numel(), accuracy(y_hat, y).item(), y.numel())

        train_loss, train_acc = metric[0] / metric[2], metric[1] / metric[2]
        net.eval(); test_metric = d2l.Accumulator(2)
        with torch.no_grad():
            for X, y in test_iter:
                X, y = X.to(device, non_blocking=True), y.to(device, non_blocking=True)
                test_metric.add(accuracy(net(X), y).item(), y.numel())
        test_acc = test_metric[0] / test_metric[1]
        scheduler.step(test_acc)
        #=== record to list==
        epochs_list.append(epoch + 1)
        train_losses.append(train_loss)
        train_accs.append(train_acc)
        test_accs.append(test_acc)
        
        animator.add(epoch + 1, (train_loss, train_acc, test_acc))
        display.clear_output(wait=True)
        display.display(animator.fig)
        animator.fig.savefig("training_curve.png", dpi=300, bbox_inches='tight')
        print(f"epoch {epoch + 1:2d}: loss {train_loss:.4f}  train_acc {train_acc:.3f}  test_acc {test_acc:.3f}")

        # Early-Stopping
        if stopper(test_acc, net):
            print("▲ Early stopping triggered. Stopping training.")
            break
    animator.axes[0].set_title("Lightweight 1D CNN Model Result")
    print(f"⏱  {timer.stop():.1f}s on {device}")
    print(f"✔ best test_acc={stopper.best_acc:.3f} — model weights saved to {stopper.best_path}")

# ===== Save CSV log =====
    with open("training_log.csv", "w", newline="") as f:
        w = csv.writer(f)
        w.writerow(["epoch", "train_loss", "train_acc", "test_acc"])
        for e, tl, ta, te in zip(epochs_list, train_losses, train_accs, test_accs):
            w.writerow([e, tl, ta, te])
    print("✔ saved metrics → training_log.csv")

 # ===== Save static curve figure =====
    plt.figure(figsize=(6.5, 4.2))
    plt.plot(epochs_list, train_losses, label="train loss")
    plt.plot(epochs_list, train_accs,  label="train acc")
    plt.plot(epochs_list, test_accs,   label="test acc")
    plt.xlabel("Epoch")
    plt.ylabel("Metric")
    plt.title("Training Curve")
    plt.legend()
    plt.grid(True, linestyle="--", alpha=0.5)
    plt.tight_layout()
    plt.savefig("training_curve.png", dpi=300, bbox_inches="tight")
    plt.show()
    print("✔ saved figure → training_curve.png")
# ===================== confusion matrix ===================== #
    print("\n▶ Generating confusion matrix for best model...")
    net.load_state_dict(torch.load(stopper.best_path, map_location=device))
    net.eval()
    y_true, y_pred = [], []
    with torch.no_grad():
        for X, y in test_iter:
            X, y = X.to(device), y.to(device)
            outputs = net(X)
            preds = outputs.argmax(1)
            y_true.extend(y.cpu().numpy())
            y_pred.extend(preds.cpu().numpy())

    cm = confusion_matrix(y_true, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["N", "V", "S", "F", "Q"])
    fig, ax = plt.subplots(figsize=(6, 6))
    disp.plot(ax=ax, cmap=plt.cm.Blues, values_format='d')
    plt.title("Confusion Matrix - ECG Classification")
    plt.savefig("confusion_matrix.png", dpi=300, bbox_inches='tight')
    plt.show()
# ===================== main ===================== #
if __name__ == "__main__":
    warnings.filterwarnings("ignore")
    download_mitbih()
    cut_beats()
    train()
```

```python
import os
import pickle, warnings, random
from pathlib import Path

import wfdb, numpy as np, neurokit2 as nk
import torch
from torch import nn
from torch.utils import data
from IPython import display
from sklearn.preprocessing import StandardScaler
from tqdm import tqdm
from d2l import torch as d2l
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from typing import List, Optional
import matplotlib.pyplot as plt
import csv

# ================= Global plotting style ================= #
plt.rcParams['font.family'] = 'Arial'
plt.rcParams['font.size'] = 11
plt.rcParams['axes.titlesize'] = 11
plt.rcParams['axes.labelsize'] = 11
plt.rcParams['xtick.labelsize'] = 11
plt.rcParams['ytick.labelsize'] = 11
plt.rcParams['legend.fontsize'] = 11

# ===================== Random Seed ===================== #
SEED = 42
random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(SEED)
    # Reproducibility settings
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False

def seed_worker(_):
    np.random.seed(SEED)
    random.seed(SEED)

# ===================== Constants ===================== #
ROOT        = Path("/Users/gaochenxi/datasets/mitbih")
RAW_DIR     = ROOT                      # wfdb data download root
MITDB_DIR   = RAW_DIR / "mitdb"         # dataset directory
TRAIN_DIR   = ROOT / "train"            # train beats .npy
TEST_DIR    = ROOT / "test"             # test beats .npy
LABELS_PKL  = ROOT / "labels.pkl"

# signal related constants
LEADS       = [0, 1]      # MLII, V5
ORIG_FS     = 360         # MIT-BIH original sampling rate
TARGET_FS   = 180         # downsample target sampling rate (ORIG_FS // 2)
SIGLEN      = 128         # fixed length after downsampling (samples)
LEFT_PAD    = 40          # left padding for R peak window (in TARGET_FS)

TEST_RECS = {str(r) for r in [101,106,108,109,112,114,115,116,118,119,
                              122,124,201,203,205,207,208,209,215,220,223,230,234]}
MIT2AAMI  = {'N':0,'L':0,'R':0,'e':0,'j':0,
             'V':1,'E':1,
             'A':2,'a':2,'J':2,'S':2,
             'F':3,
             'Q':4,'|':4,'f':4}
AAMI    = {0:'N', 1:'V', 2:'S', 3:'F', 4:'Q'}
CLASSES   = len(AAMI)             # set classes number to 5
LEAD_N        = len(LEADS)          # 2
IN_CHANNELS   = LEAD_N * 2          # time domain + frequency domain
AUG_FACTOR    = ORIG_FS / TARGET_FS # 2.0
LEFT_PAD_RAW  = int(LEFT_PAD * AUG_FACTOR)
SIGLEN_RAW    = int(SIGLEN * AUG_FACTOR)

# ===================== 1. Download ===================== #

def _list_record_headers() -> List[Path]:
    """search for .hea files under the mitdb directory """
    if not MITDB_DIR.exists():
        return []
    return sorted(MITDB_DIR.rglob("*.hea"))

def download_mitbih() -> None:
    """ensure MIT-BIH 48 records exist locally """
    hea_files = _list_record_headers()
    need_redownload = (
        len(hea_files) < 48 or
        len(list(MITDB_DIR.rglob("*.dat"))) < 48 or
        len(list(MITDB_DIR.rglob("*.hea"))) < 48 or
        len(list(MITDB_DIR.rglob("*.atr"))) < 48
    )
    if not need_redownload:
        print("MIT-BIH already downloaded, skipping")
        return

    print(f"(Re) Downloading MIT-BIH → {RAW_DIR}")
    import shutil
    # remove only the mitdb directory
    shutil.rmtree(MITDB_DIR, ignore_errors=True)
    MITDB_DIR.mkdir(parents=True, exist_ok=True)
    # wfdb create RAW_DIR/mitdb/   
    wfdb.dl_database("mitdb", dl_dir=str(RAW_DIR))
    print("Download complete")

# ===================== 2. Beat Segmentation ===================== #

def _filter(sig: np.ndarray, fs: int = ORIG_FS) -> np.ndarray:
    """0.5–40 Hz bandpass filter (neurokit2) """
    return nk.signal_filter(sig, sampling_rate=fs, lowcut=0.5, highcut=40)

def _resample(sig: np.ndarray, desired_len: int = SIGLEN) -> np.ndarray:
    """resample a 1D signal to the desired length """
    return nk.signal_resample(sig, desired_length=desired_len)

def cut_beats() -> None:
    """ beat segmentation and normalization """
    if TRAIN_DIR.exists() and any(TRAIN_DIR.glob("*.npy")) and \
       TEST_DIR.exists() and any(TEST_DIR.glob("*.npy")) and \
       LABELS_PKL.exists():
        print(".npy files already exist, skipping cut_beats()")
        return

    TRAIN_DIR.mkdir(parents=True, exist_ok=True)
    TEST_DIR.mkdir(parents=True,  exist_ok=True)

    hea_files = _list_record_headers()
    if len(hea_files) < 48:
        raise FileNotFoundError("MIT-BIH records incomplete, please check download directory")

    #  scaler（only train records）
    train_files = [p for p in hea_files if p.stem not in TEST_RECS]
    scaler = StandardScaler()
    print("Pass-1: compute mean/std on TRAIN records")
    for hea in tqdm(train_files):
        rec_base = str(hea.with_suffix(""))  # remove .hea
        sig, _ = wfdb.rdsamp(rec_base, channels=LEADS)
        sig = np.vstack([_filter(sig[:, i]) for i in range(LEAD_N)]).T  # (T, L)
        scaler.partial_fit(sig.reshape(-1, 1))

    # segment and save
    print("Pass-2: segment beats and save .npy")
    labels = {}
    for hea in tqdm(hea_files):
        rec = hea.stem
        rec_base = str(hea.with_suffix(""))
        record = wfdb.rdrecord(rec_base, channels=LEADS)
        ann    = wfdb.rdann(rec_base, "atr")

        sig = record.p_signal.astype("float32")  # (T, L)
        sig = np.vstack([_filter(sig[:, i]) for i in range(LEAD_N)]).T
        sig = scaler.transform(sig.reshape(-1, 1)).reshape(-1, LEAD_N)  # Z-score

        for idx, (r, sym) in enumerate(zip(ann.sample, ann.symbol)):
            if sym not in MIT2AAMI:
                continue
            start, end = r - LEFT_PAD_RAW, r - LEFT_PAD_RAW + SIGLEN_RAW
            if start < 0 or end > len(sig):
                continue
            beat = sig[start:end]             # (SIGLEN_RAW, L)
            beat = beat.T                     # (L, SIGLEN_RAW)
            beat = np.stack([_resample(ch, SIGLEN) for ch in beat])  # (L, SIGLEN)
            fname   = f"{rec}_{idx:05d}.npy"
            target  = TEST_DIR if rec in TEST_RECS else TRAIN_DIR
            np.save(target / fname, beat.astype("float32"))
            labels[f"{rec}_{idx:05d}"] = MIT2AAMI[sym]

    with open(LABELS_PKL, "wb") as f:
        pickle.dump(labels, f)

    print(f"Beats saved: train={len(list(TRAIN_DIR.glob('*.npy')))}, "
          f"test={len(list(TEST_DIR.glob('*.npy')))}")

# ===================== 3. Dataset ===================== #
class ECGDataset(data.Dataset):
    """time domain + frequency domain """
    def __init__(self, train: bool = True):
        self.dir   = TRAIN_DIR if train else TEST_DIR
        self.files = sorted(self.dir.glob("*.npy"))
        with open(LABELS_PKL, "rb") as f:
            self.labels = pickle.load(f)

    def __len__(self):
        return len(self.files)

    def __getitem__(self, idx):
        fp   = self.files[idx]
        x_td = np.load(fp)                        # (L, SIGLEN)

        # frequency domain： real FFT magnitude + log1p compression 
        x_fd = np.abs(np.fft.rfft(x_td, axis=1)).astype("float32")   # (L, SIGLEN//2+1)
        x_fd = np.log1p(x_fd)
        # resample frequency feature length to match time-domain length
        x_fd = np.stack([_resample(ch, SIGLEN) for ch in x_fd], axis=0)  # (L, SIGLEN)

        x    = np.concatenate([x_td, x_fd], axis=0)  # (L*2, SIGLEN)
        y    = int(self.labels[fp.stem])
        return torch.tensor(x), torch.tensor(y, dtype=torch.long)

def load_data_ecg(batch_size: int = 256):
    train_set  = ECGDataset(True)
    test_set   = ECGDataset(False)
    g = torch.Generator()
    g.manual_seed(SEED)
    pin = torch.cuda.is_available()
    num_workers = min(4, os.cpu_count() or 1)
    train_iter = data.DataLoader(train_set, batch_size, shuffle=True,  num_workers=num_workers,
                                 drop_last=True, worker_init_fn=seed_worker, generator=g,
                                 pin_memory=pin)
    test_iter  = data.DataLoader(test_set,  batch_size, shuffle=False, num_workers=num_workers,
                                 worker_init_fn=seed_worker, generator=g,
                                 pin_memory=pin)
    return train_iter, test_iter

# ===================== 4. Model ===================== #
class ECGNet(nn.Module):
    def __init__(self, in_channels: int = IN_CHANNELS, classes: int = 5, drop_p: float = 0.3):
        super().__init__()
        self.net = nn.Sequential(
            nn.Conv1d(in_channels,  32, 7, 2, 3), nn.BatchNorm1d(32),  nn.ReLU(), nn.Dropout(drop_p),
            nn.Conv1d(32,  64, 5, 2, 2),          nn.BatchNorm1d(64),  nn.ReLU(), nn.Dropout(drop_p),
            nn.Conv1d(64, 128, 3, 2, 1),          nn.BatchNorm1d(128), nn.ReLU(),
            nn.AdaptiveAvgPool1d(1), nn.Flatten(),
            nn.Dropout(drop_p),                   
            nn.Linear(128, classes)
        )

    def forward(self, x):
        return self.net(x)  # x: (B, C, T)

def init_weights(m):
    if isinstance(m, (nn.Conv1d, nn.Linear)):
        nn.init.kaiming_normal_(m.weight, nonlinearity="relu")

# ===================== 5. Early-Stopping ===================== #
class EarlyStopping:
    """Stop training when a  validation metric fails to improve for patience epochs """
    def __init__(self, patience: int = 10, verbose: bool = True):
        self.patience = patience
        self.best_acc = 0.0
        self.num_bad  = 0
        self.verbose  = verbose
        self.best_path = "best_model.pth"

    def __call__(self, acc: float, model: nn.Module):
        if acc > self.best_acc:
            self.best_acc = acc
            self.num_bad  = 0
            torch.save(model.state_dict(), self.best_path)
            if self.verbose:
                print(f"  new best test_acc={acc:.3f} — saved → {self.best_path}")
        else:
            self.num_bad += 1
            if self.verbose:
                print(f"  test_acc did not improve for {self.num_bad}/{self.patience} epochs")
        return self.num_bad >= self.patience

# ===================== 6. Train ===================== #

def accuracy(y_hat, y):
    return (y_hat.argmax(1) == y).sum()

def train(num_epochs: int = 15, batch: int = 1024, lr: float = 1e-3, device: Optional[str] = None,
          drop_p: float = 0.3, weight_decay: float = 1e-4, patience_lr: int = 2, patience_es: int = 10):
    device = device or (
        "mps"  if torch.backends.mps.is_available() else
        "cuda" if torch.cuda.is_available() else
        "cpu")

    train_iter, test_iter = load_data_ecg(batch)
    C=CLASSES
    net = ECGNet(IN_CHANNELS, C, drop_p).to(device)
    net.apply(init_weights)

    loss_fn  = nn.CrossEntropyLoss()
    opt      = torch.optim.Adam(net.parameters(), lr, weight_decay=weight_decay)
    scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(opt, mode="max", factor=0.5,
                                                           patience=patience_lr, verbose=True)
    stopper   = EarlyStopping(patience_es)
    animator  = d2l.Animator(xlabel="epoch", ylabel="metric", xlim=[1, num_epochs],
                             legend=["train loss", "train acc", "test acc"])
    timer     = d2l.Timer()

    epochs_list, train_losses, train_accs, test_accs = [], [], [], []

    for epoch in range(num_epochs):
        metric = d2l.Accumulator(3)
        net.train()
        for X, y in train_iter:
            X, y = X.to(device, non_blocking=True), y.to(device, non_blocking=True)
            y_hat = net(X)
            l = loss_fn(y_hat, y)
            opt.zero_grad(); l.backward(); opt.step()
            metric.add(l.item() * y.numel(), accuracy(y_hat, y).item(), y.numel())

        train_loss, train_acc = metric[0] / metric[2], metric[1] / metric[2]
        net.eval(); test_metric = d2l.Accumulator(2)
        with torch.no_grad():
            for X, y in test_iter:
                X, y = X.to(device, non_blocking=True), y.to(device, non_blocking=True)
                test_metric.add(accuracy(net(X), y).item(), y.numel())
        test_acc = test_metric[0] / test_metric[1]
        scheduler.step(test_acc)

        epochs_list.append(epoch + 1)
        train_losses.append(train_loss)
        train_accs.append(train_acc)
        test_accs.append(test_acc)

        animator.add(epoch + 1, (train_loss, train_acc, test_acc))
        display.clear_output(wait=True)
        display.display(animator.fig)

        print(f"epoch {epoch + 1:2d}: loss {train_loss:.4f}  train_acc {train_acc:.3f}  test_acc {test_acc:.3f}")

        # early-stopping
        if stopper(test_acc, net):
            print("Early stopping triggered.")
            break

    animator.axes[0].set_title("Lightweight 1D CNN Result")
    elapsed = timer.stop()
    print(f"Time: {elapsed:.1f}s on {device}")
    print(f"Best test_acc={stopper.best_acc:.3f} — weights saved to {stopper.best_path}")

    # ===== Save CSV log =====
    with open("training_log.csv", "w", newline="") as f:
        w = csv.writer(f)
        w.writerow(["epoch", "train_loss", "train_acc", "test_acc"])
        for e, tl, ta, va in zip(epochs_list, train_losses, train_accs, test_accs):
            w.writerow([e, tl, ta, va])
    print("Saved metrics → training_log.csv")

    # ===== Save static curve figure =====
    plt.figure(figsize=(6.5, 4.2))
    plt.plot(epochs_list, train_losses, label="train loss")
    plt.plot(epochs_list, train_accs,  label="train acc")
    plt.plot(epochs_list, test_accs,   label="test acc")
    plt.xlabel("Epoch")
    plt.ylabel("Value")
    plt.title("Training Curve")
    plt.legend()
    plt.grid(True, linestyle="--", alpha=0.5)
    plt.tight_layout()
    plt.savefig("training_curve.png", dpi=300, bbox_inches="tight")
    plt.show()
    print("Saved figure → training_curve.png")

    # ===================== Confusion matrix ===================== #
    print("\nGenerating confusion matrix for best model...")
    net.load_state_dict(torch.load(stopper.best_path, map_location=device))
    net.eval()

    def _to_int_list(arr) -> List[int]:
        a = np.asarray(arr).reshape(-1)
        return [int(x) for x in a]

    y_true, y_pred = [], []
    with torch.no_grad():
        for X, y in test_iter:
            X, y = X.to(device, non_blocking=True), y.to(device, non_blocking=True)
            outputs = net(X)
            preds = outputs.argmax(1)
            # force Python ints to avoid weird dtypes mixing in sets
            y_true.extend(_to_int_list(y.cpu().numpy()))
            y_pred.extend(_to_int_list(preds.cpu().numpy()))

    labels_used = sorted(set(y_true) | set(y_pred))
    valid_ids = set(AAMI.keys())
    unknown = [v for v in labels_used if v not in valid_ids]

    if unknown:
        from collections import Counter
        print("  Warning: unknown label ids detected in evaluation:", unknown)
        print("  y_true counts:", Counter(y_true))
        print("  y_pred counts:", Counter(y_pred))
        labels_for_cm = sorted(valid_ids & set(labels_used))
        if not labels_for_cm:
            labels_for_cm = sorted(valid_ids)
    else:
        labels_for_cm = labels_used

    cm = confusion_matrix(y_true, y_pred, labels=labels_for_cm)
    disp_labels = [AAMI.get(i, str(i)) for i in labels_for_cm]

    fig, ax = plt.subplots(figsize=(6, 6))
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=disp_labels)
    disp.plot(ax=ax, cmap=plt.cm.Blues, values_format='d')
    plt.title("Confusion Matrix - ECG Classification")
    plt.tight_layout()
    plt.savefig("confusion_matrix.png", dpi=300, bbox_inches='tight')
    plt.show()
    print("Saved figure → confusion_matrix.png")
# ===================== main ===================== #
if __name__ == "__main__":
    warnings.filterwarnings("ignore")
    download_mitbih()
    cut_beats()
    train()
```

```python
import os
import pickle, warnings, random
from pathlib import Path

import wfdb, numpy as np, neurokit2 as nk
import torch
from torch import nn
from torch.utils import data
from IPython import display
from sklearn.preprocessing import StandardScaler
from tqdm import tqdm
from d2l import torch as d2l
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
import csv

# ================= Global plotting style ================= #
plt.rcParams['font.family'] = 'Arial'
plt.rcParams['font.size'] = 11
plt.rcParams['axes.titlesize'] = 11
plt.rcParams['axes.labelsize'] = 11
plt.rcParams['xtick.labelsize'] = 11
plt.rcParams['ytick.labelsize'] = 11
plt.rcParams['legend.fontsize'] = 11

# ===================== Random Seed ===================== #
SEED = 42
random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(SEED)
    # Reproducibility settings
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False

def seed_worker(_):
    np.random.seed(SEED)
    random.seed(SEED)

# ===================== Constants ===================== #
ROOT        = Path("/Users/gaochenxi/datasets/mitbih")
RAW_DIR     = ROOT                      # WFDB data download root
MITDB_DIR   = RAW_DIR / "mitdb"         # Dataset directory
TRAIN_DIR   = ROOT / "train"            # Train beats (.npy)
TEST_DIR    = ROOT / "test"             # Test beats (.npy)
LABELS_PKL  = ROOT / "labels.pkl"

# Signal-related constants
LEADS       = [0, 1]      # MLII, V5
ORIG_FS     = 360         # MIT-BIH original sampling rate
TARGET_FS   = 180         # Downsample target sampling rate (ORIG_FS // 2)
SIGLEN      = 128         # Fixed length after downsampling (samples)
LEFT_PAD    = 40          # Left padding for R-peak window (in TARGET_FS)

TEST_RECS = {str(r) for r in [101,106,108,109,112,114,115,116,118,119,
                              122,124,201,203,205,207,208,209,215,220,223,230,234]}
MIT2AAMI  = {'N':0,'L':0,'R':0,'e':0,'j':0,
             'V':1,'E':1,
             'A':2,'a':2,'J':2,'S':2,
             'F':3,
             'Q':4,'|':4,'f':4}
AAMI      = {0:'N', 1:'V', 2:'S', 3:'F', 4:'Q'}
CLASSES   = len(AAMI)     # Fixed number of classes: 5
LEAD_N        = len(LEADS)          # 2
IN_CHANNELS   = LEAD_N * 2          # Time domain + frequency domain
AUG_FACTOR    = ORIG_FS / TARGET_FS # 2.0
LEFT_PAD_RAW  = int(LEFT_PAD * AUG_FACTOR)
SIGLEN_RAW    = int(SIGLEN * AUG_FACTOR)

# ===================== 1. Download ===================== #

def _list_record_headers() -> list[Path]:
    """Search for .hea files under the mitdb directory."""
    if not MITDB_DIR.exists():
        return []
    return sorted(MITDB_DIR.rglob("*.hea"))

def download_mitbih() -> None:
    """Ensure the 48 MIT-BIH records exist locally."""
    hea_files = _list_record_headers()
    need_redownload = (
        len(hea_files) < 48 or
        len(list(MITDB_DIR.rglob("*.dat"))) < 48 or
        len(list(MITDB_DIR.rglob("*.hea"))) < 48 or
        len(list(MITDB_DIR.rglob("*.atr"))) < 48
    )
    if not need_redownload:
        print("MIT-BIH already downloaded, skipping")
        return

    print(f"(Re) Downloading MIT-BIH → {RAW_DIR}")
    import shutil
    # Remove only the mitdb directory
    shutil.rmtree(MITDB_DIR, ignore_errors=True)
    MITDB_DIR.mkdir(parents=True, exist_ok=True)
    # WFDB will create RAW_DIR/mitdb/
    wfdb.dl_database("mitdb", dl_dir=str(RAW_DIR))
    print("Download complete")

# ===================== 2. Beat Segmentation ===================== #

def _filter(sig: np.ndarray, fs: int = ORIG_FS) -> np.ndarray:
    """0.5–40 Hz band-pass filter (neurokit2)."""
    return nk.signal_filter(sig, sampling_rate=fs, lowcut=0.5, highcut=40)

def _resample(sig: np.ndarray, desired_len: int = SIGLEN) -> np.ndarray:
    """Resample a 1D signal to the desired length."""
    return nk.signal_resample(sig, desired_length=desired_len)

def cut_beats() -> None:
    """Beat segmentation and normalization.

    To avoid data leakage, the StandardScaler is fitted on training records only.
    """
    if TRAIN_DIR.exists() and any(TRAIN_DIR.glob("*.npy")) and \
       TEST_DIR.exists() and any(TEST_DIR.glob("*.npy")) and \
       LABELS_PKL.exists():
        print(".npy files already exist, skipping cut_beats()")
        return

    TRAIN_DIR.mkdir(parents=True, exist_ok=True)
    TEST_DIR.mkdir(parents=True,  exist_ok=True)

    hea_files = _list_record_headers()
    if len(hea_files) < 48:
        raise FileNotFoundError("MIT-BIH records incomplete; please check the download directory")

    # Fit scaler using training records only
    train_files = [p for p in hea_files if p.stem not in TEST_RECS]
    scaler = StandardScaler()
    print("Pass-1: compute mean/std on TRAIN records")
    for hea in tqdm(train_files):
        rec_base = str(hea.with_suffix(""))  # Remove .hea
        sig, _ = wfdb.rdsamp(rec_base, channels=LEADS)
        sig = np.vstack([_filter(sig[:, i]) for i in range(LEAD_N)]).T  # (T, L)
        scaler.partial_fit(sig.reshape(-1, 1))

    # Segment and save beats
    print("Pass-2: segment beats and save .npy")
    labels = {}
    for hea in tqdm(hea_files):
        rec = hea.stem
        rec_base = str(hea.with_suffix(""))
        record = wfdb.rdrecord(rec_base, channels=LEADS)
        ann    = wfdb.rdann(rec_base, "atr")

        sig = record.p_signal.astype("float32")  # (T, L)
        sig = np.vstack([_filter(sig[:, i]) for i in range(LEAD_N)]).T
        sig = scaler.transform(sig.reshape(-1, 1)).reshape(-1, LEAD_N)  # Z-score

        for idx, (r, sym) in enumerate(zip(ann.sample, ann.symbol)):
            if sym not in MIT2AAMI:
                continue
            start, end = r - LEFT_PAD_RAW, r - LEFT_PAD_RAW + SIGLEN_RAW
            if start < 0 or end > len(sig):
                continue
            beat = sig[start:end]             # (SIGLEN_RAW, L)
            beat = beat.T                     # (L, SIGLEN_RAW)
            beat = np.stack([_resample(ch, SIGLEN) for ch in beat])  # (L, SIGLEN)
            fname   = f"{rec}_{idx:05d}.npy"
            target  = TEST_DIR if rec in TEST_RECS else TRAIN_DIR
            np.save(target / fname, beat.astype("float32"))
            labels[f"{rec}_{idx:05d}"] = MIT2AAMI[sym]

    with open(LABELS_PKL, "wb") as f:
        pickle.dump(labels, f)

    print(f"Beats saved: train={len(list(TRAIN_DIR.glob('*.npy')))}, "
          f"test={len(list(TEST_DIR.glob('*.npy')))}")

# ===================== 3. Dataset ===================== #
class ECGDataset(data.Dataset):
    """Return time-domain + frequency-domain channels for each beat."""
    def __init__(self, train: bool = True):
        self.dir   = TRAIN_DIR if train else TEST_DIR
        self.files = sorted(self.dir.glob("*.npy"))
        with open(LABELS_PKL, "rb") as f:
            self.labels = pickle.load(f)

    def __len__(self):
        return len(self.files)

    def __getitem__(self, idx):
        fp   = self.files[idx]
        x_td = np.load(fp)  # (L, SIGLEN)

        # Frequency domain: real-FFT magnitude + log1p compression
        x_fd = np.abs(np.fft.rfft(x_td, axis=1)).astype("float32")   # (L, SIGLEN//2+1)
        x_fd = np.log1p(x_fd)
        # Resample frequency feature length to match the time-domain length
        x_fd = np.stack([_resample(ch, SIGLEN) for ch in x_fd], axis=0)  # (L, SIGLEN)

        x    = np.concatenate([x_td, x_fd], axis=0)  # (L*2, SIGLEN)
        y    = int(self.labels[fp.stem])
        return torch.tensor(x), torch.tensor(y, dtype=torch.long)

def load_data_ecg(batch_size: int = 256):
    """Create DataLoaders for training and testing."""
    train_set  = ECGDataset(True)
    test_set   = ECGDataset(False)

    g = torch.Generator(); g.manual_seed(SEED)
    pin = torch.cuda.is_available()

    train_iter = data.DataLoader(train_set, batch_size, shuffle=True,
                                 num_workers=0, drop_last=True,
                                 worker_init_fn=seed_worker, generator=g,
                                 pin_memory=pin)
    test_iter  = data.DataLoader(test_set,  batch_size, shuffle=False,
                                 num_workers=0,
                                 worker_init_fn=seed_worker, generator=g,
                                 pin_memory=pin)
    return train_iter, test_iter

# ===================== 4. Model ===================== #
class ECGNet(nn.Module):
    def __init__(self, in_channels: int = IN_CHANNELS, classes: int = 5, drop_p: float = 0.3):
        super().__init__()
        self.net = nn.Sequential(
            nn.Conv1d(in_channels,  32, 7, 2, 3), nn.BatchNorm1d(32),  nn.ReLU(), nn.Dropout(drop_p),
            nn.Conv1d(32,  64, 5, 2, 2),          nn.BatchNorm1d(64),  nn.ReLU(), nn.Dropout(drop_p),
            nn.Conv1d(64, 128, 3, 2, 1),          nn.BatchNorm1d(128), nn.ReLU(),
            nn.AdaptiveAvgPool1d(1), nn.Flatten(),
            nn.Dropout(drop_p),
            nn.Linear(128, classes)
        )

    def forward(self, x):
        return self.net(x)  # x: (B, C, T)

def init_weights(m):
    if isinstance(m, (nn.Conv1d, nn.Linear)):
        nn.init.kaiming_normal_(m.weight, nonlinearity="relu")

# ===================== 5. Early-Stopping ===================== #
class EarlyStopping:
    """Stop training when the test metric fails to improve for `patience` epochs."""
    def __init__(self, patience: int = 10, verbose: bool = True):
        self.patience = patience
        self.best_acc = 0.0
        self.num_bad  = 0
        self.verbose  = verbose
        self.best_path = "best_model.pth"

    def __call__(self, acc: float, model: nn.Module):
        if acc > self.best_acc:
            self.best_acc = acc
            self.num_bad  = 0
            torch.save(model.state_dict(), self.best_path)
            if self.verbose:
                print(f"  new best test_acc={acc:.3f} — saved → {self.best_path}")
        else:
            self.num_bad += 1
            if self.verbose:
                print(f"  test_acc did not improve for {self.num_bad}/{self.patience} epochs")
        return self.num_bad >= self.patience

# ===================== 6. Train ===================== #

def accuracy(y_hat, y):
    return (y_hat.argmax(1) == y).sum()

def train(num_epochs: int = 15, batch: int = 1024, lr: float = 1e-3, device: str | None = None,
          drop_p: float = 0.3, weight_decay: float = 1e-4, patience_lr: int = 2, patience_es: int = 10):
    device = device or (
        "mps"  if torch.backends.mps.is_available() else
        "cuda" if torch.cuda.is_available() else
        "cpu")

    train_iter, test_iter = load_data_ecg(batch)
    C = CLASSES
    net = ECGNet(IN_CHANNELS, C, drop_p).to(device)
    net.apply(init_weights)

    loss_fn  = nn.CrossEntropyLoss()
    opt      = torch.optim.Adam(net.parameters(), lr, weight_decay=weight_decay)
    scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(opt, mode="max", factor=0.5,
                                                           patience=patience_lr, verbose=True)
    stopper   = EarlyStopping(patience_es)
    animator  = d2l.Animator(xlabel="epoch", ylabel="metric", xlim=[1, num_epochs],
                             legend=["train loss", "train acc", "test acc"])
    timer     = d2l.Timer()

    epochs_list, train_losses, train_accs, test_accs = [], [], [], []

    for epoch in range(num_epochs):
        metric = d2l.Accumulator(3)
        net.train()
        for X, y in train_iter:
            X, y = X.to(device, non_blocking=True), y.to(device, non_blocking=True)
            y_hat = net(X)
            l = loss_fn(y_hat, y)
            opt.zero_grad(); l.backward(); opt.step()
            metric.add(l.item() * y.numel(), accuracy(y_hat, y).item(), y.numel())

        train_loss, train_acc = metric[0] / metric[2], metric[1] / metric[2]
        net.eval(); test_metric = d2l.Accumulator(2)
        with torch.no_grad():
            for X, y in test_iter:
                X, y = X.to(device, non_blocking=True), y.to(device, non_blocking=True)
                test_metric.add(accuracy(net(X), y).item(), y.numel())
        test_acc = test_metric[0] / test_metric[1]
        scheduler.step(test_acc)

        epochs_list.append(epoch + 1)
        train_losses.append(train_loss)
        train_accs.append(train_acc)
        test_accs.append(test_acc)

        animator.add(epoch + 1, (train_loss, train_acc, test_acc))
        display.clear_output(wait=True)
        display.display(animator.fig)

        print(f"epoch {epoch + 1:2d}: loss {train_loss:.4f}  train_acc {train_acc:.3f}  test_acc {test_acc:.3f}")

        # Early stopping
        if stopper(test_acc, net):
            print("Early stopping triggered.")
            break

    animator.axes[0].set_title("Lightweight 1D CNN Result")
    elapsed = timer.stop()
    print(f"Time: {elapsed:.1f}s on {device}")
    print(f"Best test_acc={stopper.best_acc:.3f} — weights saved to {stopper.best_path}")

    # ===== Save CSV log =====
    with open("training_log.csv", "w", newline="") as f:
        w = csv.writer(f)
        w.writerow(["epoch", "train_loss", "train_acc", "test_acc"])
        for e, tl, ta, va in zip(epochs_list, train_losses, train_accs, test_accs):
            w.writerow([e, tl, ta, va])
    print("Saved metrics → training_log.csv")

    # ===== Save static curve figure =====
    plt.figure(figsize=(6.5, 4.2))
    plt.plot(epochs_list, train_losses, label="train loss")
    plt.plot(epochs_list, train_accs,  label="train acc")
    plt.plot(epochs_list, test_accs,   label="test acc")
    plt.xlabel("Epoch")
    plt.ylabel("Value")
    plt.title("Training Curve")
    plt.legend()
    plt.grid(True, linestyle="--", alpha=0.5)
    plt.tight_layout()
    plt.savefig("training_curve.png", dpi=300, bbox_inches="tight")
    plt.show()
    print("Saved figure → training_curve.png")

    # ===================== Confusion matrix ===================== #
    print("\nGenerating confusion matrix for the best model...")
    net.load_state_dict(torch.load(stopper.best_path, map_location=device))
    net.eval()

    def _to_int_list(arr) -> list[int]:
        a = np.asarray(arr).reshape(-1)
        return [int(x) for x in a]

    y_true, y_pred = [], []
    with torch.no_grad():
        for X, y in test_iter:
            X, y = X.to(device, non_blocking=True), y.to(device, non_blocking=True)
            outputs = net(X)
            preds = outputs.argmax(1)
            # Force Python ints to avoid weird dtypes mixing in sets
            y_true.extend(_to_int_list(y.cpu().numpy()))
            y_pred.extend(_to_int_list(preds.cpu().numpy()))

    labels_used = sorted(set(y_true) | set(y_pred))
    valid_ids = set(AAMI.keys())
    unknown = [v for v in labels_used if v not in valid_ids]

    if unknown:
        from collections import Counter
        print("  Warning: unknown label ids detected in evaluation:", unknown)
        print("  y_true counts:", Counter(y_true))
        print("  y_pred counts:", Counter(y_pred))
        labels_for_cm = sorted(valid_ids & set(labels_used))
        if not labels_for_cm:
            labels_for_cm = sorted(valid_ids)
    else:
        labels_for_cm = labels_used

    cm = confusion_matrix(y_true, y_pred, labels=labels_for_cm)
    disp_labels = [AAMI.get(i, str(i)) for i in labels_for_cm]

    fig, ax = plt.subplots(figsize=(6, 6))
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=disp_labels)
    disp.plot(ax=ax, cmap=plt.cm.Blues, values_format='d')
    plt.title("Confusion Matrix - ECG Classification")
    plt.tight_layout()
    plt.savefig("confusion_matrix.png", dpi=300, bbox_inches='tight')
    plt.show()
    print("Saved figure → confusion_matrix.png")

# ===================== main ===================== #
if __name__ == "__main__":
    warnings.filterwarnings("ignore")
    download_mitbih()
    cut_beats()
    train()
```

```python
import pickle, warnings, random
from pathlib import Path

import wfdb, numpy as np, neurokit2 as nk
import torch
from torch import nn
from torch.utils import data
from IPython import display
from sklearn.preprocessing import StandardScaler
from tqdm import tqdm
from d2l import torch as d2l
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
import csv

# ================= Global plotting style ================= #
plt.rcParams['font.family'] = 'Arial'
plt.rcParams['font.size'] = 11
plt.rcParams['axes.titlesize'] = 11
plt.rcParams['axes.labelsize'] = 11
plt.rcParams['xtick.labelsize'] = 11
plt.rcParams['ytick.labelsize'] = 11
plt.rcParams['legend.fontsize'] = 11

# ===================== Random Seed ===================== #
SEED = 42
random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(SEED)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False

def seed_worker(_):
    np.random.seed(SEED)
    random.seed(SEED)

# ===================== Constants ===================== #
ROOT        = Path("/Users/gaochenxi/datasets/mitbih")
RAW_DIR     = ROOT                      # wfdb data download 
MITDB_DIR   = RAW_DIR / "mitdb"         # dataset directory
TRAIN_DIR   = ROOT / "train"            # train beats .npy
TEST_DIR    = ROOT / "test"             # test beats .npy
LABELS_PKL  = ROOT / "labels.pkl"

# 信号相关
LEADS       = [0, 1]      # MLII, V5
ORIG_FS     = 360         # MIT-BIH original sample rate
TARGET_FS   = 180         # Downsample traget sampling rate (ORIG_FS // 2)
SIGLEN      = 128         # Fixed length after downsampling (samples)
LEFT_PAD    = 40          # Left padding for R peak window (in TARGET_FS)

TEST_RECS = {str(r) for r in [101,106,108,109,112,114,115,116,118,119,
                              122,124,201,203,205,207,208,209,215,220,223,230,234]}
MIT2AAMI  = {'N':0,'L':0,'R':0,'e':0,'j':0,
             'V':1,'E':1,
             'A':2,'a':2,'J':2,'S':2,
             'F':3,
             'Q':4,'|':4,'f':4}
ID2LBL    = {0:'N', 1:'V', 2:'S', 3:'F', 4:'Q'}
CLASSES   = len(ID2LBL)             # set classes number to 5
LEAD_N        = len(LEADS)          # 2
IN_CHANNELS   = LEAD_N * 2          # time domain + frequency domain
AUG_FACTOR    = ORIG_FS / TARGET_FS # 2.0
LEFT_PAD_RAW  = int(LEFT_PAD * AUG_FACTOR)
SIGLEN_RAW    = int(SIGLEN * AUG_FACTOR)

# ===================== 1. Download ===================== #

def _list_record_headers() -> list[Path]:
    """search .hea files under mitdb directory """
    if not MITDB_DIR.exists():
        return []
    return sorted(MITDB_DIR.rglob("*.hea"))

def download_mitbih() -> None:
    """ensure MIT-BIH 48 records exist locally """
    hea_files = _list_record_headers()
    need_redownload = (
        len(hea_files) < 48 or
        len(list(MITDB_DIR.rglob("*.dat"))) < 48 or
        len(list(MITDB_DIR.rglob("*.hea"))) < 48 or
        len(list(MITDB_DIR.rglob("*.atr"))) < 48
    )
    if not need_redownload:
        print("MIT-BIH already downloaded, skipping")
        return

    print(f"(Re) Downloading MIT-BIH → {RAW_DIR}")
    import shutil
    # delete mitdb directory
    shutil.rmtree(MITDB_DIR, ignore_errors=True)
    MITDB_DIR.mkdir(parents=True, exist_ok=True)
    # wfdb construct mitdb/ under RAW_DIR  
    wfdb.dl_database("mitdb", dl_dir=str(RAW_DIR))
    print("Download complete")

# ===================== 2. Beat Segmentation ===================== #

def _filter(sig: np.ndarray, fs: int = ORIG_FS) -> np.ndarray:
    """0.5–40 Hz bandpass filter（neurokit2）"""
    return nk.signal_filter(sig, sampling_rate=fs, lowcut=0.5, highcut=40)

def _resample(sig: np.ndarray, desired_len: int = SIGLEN) -> np.ndarray:
    """Resample signal to desired_len samples """
    return nk.signal_resample(sig, desired_length=desired_len)

def cut_beats() -> None:
    """ Beat segmentation and normalization """
    if TRAIN_DIR.exists() and any(TRAIN_DIR.glob("*.npy")) and \
       TEST_DIR.exists() and any(TEST_DIR.glob("*.npy")) and \
       LABELS_PKL.exists():
        print(".npy files already exist, skipping cut_beats()")
        return

    TRAIN_DIR.mkdir(parents=True, exist_ok=True)
    TEST_DIR.mkdir(parents=True,  exist_ok=True)

    hea_files = _list_record_headers()
    if len(hea_files) < 48:
        raise FileNotFoundError("MIT-BIH records incomplete, please check download directory")

    #  scaler（only train records）
    train_files = [p for p in hea_files if p.stem not in TEST_RECS]
    scaler = StandardScaler()
    print("Pass-1: compute mean/std on TRAIN records")
    for hea in tqdm(train_files):
        rec_base = str(hea.with_suffix(""))  # remove .hea
        sig, _ = wfdb.rdsamp(rec_base, channels=LEADS)
        sig = np.vstack([_filter(sig[:, i]) for i in range(LEAD_N)]).T  # (T, L)
        scaler.partial_fit(sig.reshape(-1, 1))

    # cut and save
    print("Pass-2: segment beats and save .npy")
    labels = {}
    for hea in tqdm(hea_files):
        rec = hea.stem
        rec_base = str(hea.with_suffix(""))
        record = wfdb.rdrecord(rec_base, channels=LEADS)
        ann    = wfdb.rdann(rec_base, "atr")

        sig = record.p_signal.astype("float32")  # (T, L)
        sig = np.vstack([_filter(sig[:, i]) for i in range(LEAD_N)]).T
        sig = scaler.transform(sig.reshape(-1, 1)).reshape(-1, LEAD_N)  # Z-score

        for idx, (r, sym) in enumerate(zip(ann.sample, ann.symbol)):
            if sym not in MIT2AAMI:
                continue
            start, end = r - LEFT_PAD_RAW, r - LEFT_PAD_RAW + SIGLEN_RAW
            if start < 0 or end > len(sig):
                continue
            beat = sig[start:end]             # (SIGLEN_RAW, L)
            beat = beat.T                     # (L, SIGLEN_RAW)
            beat = np.stack([_resample(ch, SIGLEN) for ch in beat])  # (L, SIGLEN)
            fname   = f"{rec}_{idx:05d}.npy"
            target  = TEST_DIR if rec in TEST_RECS else TRAIN_DIR
            np.save(target / fname, beat.astype("float32"))
            labels[f"{rec}_{idx:05d}"] = MIT2AAMI[sym]

    with open(LABELS_PKL, "wb") as f:
        pickle.dump(labels, f)

    print(f"Beats saved: train={len(list(TRAIN_DIR.glob('*.npy')))}, "
          f"test={len(list(TEST_DIR.glob('*.npy')))}")

# ===================== 3. Dataset ===================== #
class ECGDataset(data.Dataset):
    """time domain + frequency domain """
    def __init__(self, train: bool = True):
        self.dir   = TRAIN_DIR if train else TEST_DIR
        self.files = sorted(self.dir.glob("*.npy"))
        with open(LABELS_PKL, "rb") as f:
            self.labels = pickle.load(f)

    def __len__(self):
        return len(self.files)

    def __getitem__(self, idx):
        fp   = self.files[idx]
        x_td = np.load(fp)                        # (L, SIGLEN)

        # frequency domain： rFFT + log1p 
        x_fd = np.abs(np.fft.rfft(x_td, axis=1)).astype("float32")   # (L, SIGLEN//2+1)
        x_fd = np.log1p(x_fd)
        # resample signals to the length same with the time domain
        x_fd = np.stack([_resample(ch, SIGLEN) for ch in x_fd], axis=0)  # (L, SIGLEN)

        x    = np.concatenate([x_td, x_fd], axis=0)  # (L*2, SIGLEN)
        y    = int(self.labels[fp.stem])
        return torch.tensor(x), torch.tensor(y, dtype=torch.long)

def load_data_ecg(batch_size: int = 256):
    train_set  = ECGDataset(True)
    test_set   = ECGDataset(False)
    g = torch.Generator()
    g.manual_seed(SEED)
    pin = torch.cuda.is_available()
    train_iter = data.DataLoader(train_set, batch_size, shuffle=True,  num_workers=0,
                                 drop_last=True, worker_init_fn=seed_worker, generator=g,
                                 pin_memory=pin)
    test_iter  = data.DataLoader(test_set,  batch_size, shuffle=False, num_workers=0,
                                 worker_init_fn=seed_worker, generator=g,
                                 pin_memory=pin)
    return train_iter, test_iter

# ===================== 4. Model ===================== #
class ECGNet(nn.Module):
    def __init__(self, in_channels: int = IN_CHANNELS, classes: int = 5, drop_p: float = 0.3):
        super().__init__()
        self.net = nn.Sequential(
            nn.Conv1d(in_channels,  32, 7, 2, 3), nn.BatchNorm1d(32),  nn.ReLU(), nn.Dropout(drop_p),
            nn.Conv1d(32,  64, 5, 2, 2),          nn.BatchNorm1d(64),  nn.ReLU(), nn.Dropout(drop_p),
            nn.Conv1d(64, 128, 3, 2, 1),          nn.BatchNorm1d(128), nn.ReLU(),
            nn.AdaptiveAvgPool1d(1), nn.Flatten(),
            nn.Dropout(drop_p),                   
            nn.Linear(128, classes)
        )

    def forward(self, x):
        return self.net(x)  # x: (B, C, T)

def init_weights(m):
    if isinstance(m, (nn.Conv1d, nn.Linear)):
        nn.init.kaiming_normal_(m.weight, nonlinearity="relu")

# ===================== 5. Early-Stopping ===================== #
class EarlyStopping:
    """Stop training when a  validation metric fails to improve for patience epochs """
    def __init__(self, patience: int = 10, verbose: bool = True):
        self.patience = patience
        self.best_acc = 0.0
        self.num_bad  = 0
        self.verbose  = verbose
        self.best_path = "best_model.pth"

    def __call__(self, acc: float, model: nn.Module):
        if acc > self.best_acc:
            self.best_acc = acc
            self.num_bad  = 0
            torch.save(model.state_dict(), self.best_path)
            if self.verbose:
                print(f"  new best test_acc={acc:.3f} — saved → {self.best_path}")
        else:
            self.num_bad += 1
            if self.verbose:
                print(f"  test_acc did not improve for {self.num_bad}/{self.patience} epochs")
        return self.num_bad >= self.patience

# ===================== 6. Train ===================== #

def accuracy(y_hat, y):
    return (y_hat.argmax(1) == y).sum()

def train(num_epochs: int = 15, batch: int = 1024, lr: float = 1e-3, device: str | None = None,
          drop_p: float = 0.3, weight_decay: float = 1e-4, patience_lr: int = 2, patience_es: int = 10):
    device = device or (
        "mps"  if torch.backends.mps.is_available() else
        "cuda" if torch.cuda.is_available() else
        "cpu")

    train_iter, test_iter, _ = load_data_ecg(batch)
    C=CLASSES
    net = ECGNet(IN_CHANNELS, C, drop_p).to(device)
    net.apply(init_weights)

    loss_fn  = nn.CrossEntropyLoss()
    opt      = torch.optim.Adam(net.parameters(), lr, weight_decay=weight_decay)
    scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(opt, mode="max", factor=0.5,
                                                           patience=patience_lr, verbose=True)
    stopper   = EarlyStopping(patience_es)
    animator  = d2l.Animator(xlabel="epoch", ylabel="metric", xlim=[1, num_epochs],
                             legend=["train loss", "train acc", "test acc"])
    timer     = d2l.Timer()

    epochs_list, train_losses, train_accs, test_accs = [], [], [], []

    for epoch in range(num_epochs):
        metric = d2l.Accumulator(3)
        net.train()
        for X, y in train_iter:
            X, y = X.to(device, non_blocking=True), y.to(device, non_blocking=True)
            y_hat = net(X)
            l = loss_fn(y_hat, y)
            opt.zero_grad(); l.backward(); opt.step()
            metric.add(l.item() * y.numel(), accuracy(y_hat, y).item(), y.numel())

        train_loss, train_acc = metric[0] / metric[2], metric[1] / metric[2]
        net.eval(); test_metric = d2l.Accumulator(2)
        with torch.no_grad():
            for X, y in test_iter:
                X, y = X.to(device, non_blocking=True), y.to(device, non_blocking=True)
                test_metric.add(accuracy(net(X), y).item(), y.numel())
        test_acc = test_metric[0] / test_metric[1]
        scheduler.step(test_acc)

        epochs_list.append(epoch + 1)
        train_losses.append(train_loss)
        train_accs.append(train_acc)
        test_accs.append(test_acc)

        animator.add(epoch + 1, (train_loss, train_acc, test_acc))
        display.clear_output(wait=True)
        display.display(animator.fig)

        print(f"epoch {epoch + 1:2d}: loss {train_loss:.4f}  train_acc {train_acc:.3f}  test_acc {test_acc:.3f}")

        # Early-Stopping
        if stopper(test_acc, net):
            print("Early stopping triggered.")
            break

    animator.axes[0].set_title("Lightweight 1D CNN Result")
    elapsed = timer.stop()
    print(f"Time: {elapsed:.1f}s on {device}")
    print(f"Best test_acc={stopper.best_acc:.3f} — weights saved to {stopper.best_path}")

    # ===== Save CSV log =====
    with open("training_log.csv", "w", newline="") as f:
        w = csv.writer(f)
        w.writerow(["epoch", "train_loss", "train_acc", "test_acc"])
        for e, tl, ta, va in zip(epochs_list, train_losses, train_accs, test_accs):
            w.writerow([e, tl, ta, va])
    print("Saved metrics → training_log.csv")

    # ===== Save static curve figure =====
    plt.figure(figsize=(6.5, 4.2))
    plt.plot(epochs_list, train_losses, label="train loss")
    plt.plot(epochs_list, train_accs,  label="train acc")
    plt.plot(epochs_list, test_accs,   label="test acc")
    plt.xlabel("Epoch")
    plt.ylabel("Value")
    plt.title("Training Curve")
    plt.legend()
    plt.grid(True, linestyle="--", alpha=0.5)
    plt.tight_layout()
    plt.savefig("training_curve.png", dpi=300, bbox_inches="tight")
    plt.show()
    print("Saved figure → training_curve.png")

    # ===================== Confusion matrix ===================== #
    print("\nGenerating confusion matrix for best model...")
    net.load_state_dict(torch.load(stopper.best_path, map_location=device))
    net.eval()

    def _to_int_list(arr) -> list[int]:
        a = np.asarray(arr).reshape(-1)
        return [int(x) for x in a]

    y_true, y_pred = [], []
    with torch.no_grad():
        for X, y in test_iter:
            X, y = X.to(device, non_blocking=True), y.to(device, non_blocking=True)
            outputs = net(X)
            preds = outputs.argmax(1)
            # int
            y_true.extend(_to_int_list(y.cpu().numpy()))
            y_pred.extend(_to_int_list(preds.cpu().numpy()))

    labels_used = sorted(set(y_true) | set(y_pred))
    valid_ids = set(ID2LBL.keys())
    unknown = [v for v in labels_used if v not in valid_ids]

    if unknown:
        from collections import Counter
        print("  Warning: unknown label ids detected in evaluation:", unknown)
        print("  y_true counts:", Counter(y_true))
        print("  y_pred counts:", Counter(y_pred))
        labels_for_cm = sorted(valid_ids & set(labels_used))
        if not labels_for_cm:
            labels_for_cm = sorted(valid_ids)
    else:
        labels_for_cm = labels_used

    cm = confusion_matrix(y_true, y_pred, labels=labels_for_cm)
    disp_labels = [ID2LBL.get(i, str(i)) for i in labels_for_cm]

    fig, ax = plt.subplots(figsize=(6, 6))
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=disp_labels)
    disp.plot(ax=ax, cmap=plt.cm.Blues, values_format='d')
    plt.title("Confusion Matrix - ECG Classification")
    plt.tight_layout()
    plt.savefig("confusion_matrix.png", dpi=300, bbox_inches='tight')
    plt.show()
    print("Saved figure → confusion_matrix.png")
# ===================== main ===================== #
if __name__ == "__main__":
    warnings.filterwarnings("ignore")
    download_mitbih()
    cut_beats()
    train()
```
